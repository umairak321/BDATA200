<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why is Data Visualization Important?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Turning Complex Data into Clear Insights</em></p>

  <h3 style="color: #333;">Why It Matters</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Transforms complex data into clear insights — humans process visuals <strong>60,000x faster</strong> than text</li>
    <li>Reveals patterns, trends, and outliers hidden in raw numbers</li>
    <li>Enables faster, data-driven decisions across all fields</li>
  </ul>

  <h3 style="color: #333;">Main Challenges</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Choosing the right chart type for the data and audience</li>
    <li>Avoiding misleading visuals — truncated axes, cherry-picked data, 3D distortion</li>
    <li>Balancing simplicity and completeness — too much detail overwhelms, too little misleads</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Science of Visual Processing</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Why Our Brains Are Wired for Visuals</em></p>

  <h3 style="color: #333;">John Medina (Brain Rules)</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>People remember <strong>~10%</strong> of text after 3 days</li>
    <li>But <strong>~65%</strong> when paired with relevant images</li>
  </ul>

  <h3 style="color: #333;">Cognitive Science & Neuroscience</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>The visual system processes information in <strong>parallel</strong> (not sequential like text)</li>
    <li>A large portion of the brain is dedicated to vision</li>
    <li>Pattern recognition in images is extremely fast — <strong>milliseconds</strong> range</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Strategies for Presenting to Audiences</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>How to Communicate Data Findings Effectively</em></p>

  <h3 style="color: #333;">Know Your Audience</h3>
  <p style="font-size: 1.02em; color: #444;">Experts want depth; general audiences want the "so what?"</p>

  <h3 style="color: #333;">Lead with the Surprise, Not the Method</h3>
  <p style="font-size: 1.02em; color: #444;">Say <em>"Drug dealers earn <span>$</span>3.30/hr"</em> not <em>"We ran a regression on gang finances"</em></p>

  <h3 style="color: #333;">Use Contrast & Comparison</h3>
  <p style="font-size: 1.02em; color: #444;"><em>"Expected 50%, actual 75%"</em> is more powerful than <em>"win rate was 75%"</em></p>

  <h3 style="color: #333;">Make It Personal & Relatable</h3>
  <p style="font-size: 1.02em; color: #444;"><em>"Your real estate agent keeps their home on the market 10 days longer than yours"</em></p>

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Sumo Wrestlers: Honor... or Cheating?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When Incentives Override Tradition</em></p>

  <h3 style="color: #333;">The Setup</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Sumo is Japan's most sacred sport, built on centuries of honor</li>
    <li>Wrestlers with <strong>7-7 records</strong> on the final day desperately need 1 more win to keep their ranking</li>
    <li>Their opponent at <strong>8-6</strong> has nothing to lose</li>
  </ul>

  <h3 style="color: #333;">What the Data Shows</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>The 7-7 wrestler wins <strong>~75%</strong> of the time (expected: ~50%)</li>
    <li>In rematches, the pattern mysteriously <strong>reverses</strong></li>
    <li>When media investigated, the suspicious wins suddenly <strong>disappeared</strong></li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> Incentives shape behavior — even in the most honor-bound traditions.</p>
  </blockquote>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Did Crime Drop in the 1990s?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Most Controversial Finding in Freakonomics</em></p>

  <h3 style="color: #333;">Everyone Said It Was Because Of:</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Better policing strategies</li>
    <li>Harsher prison sentences</li>
    <li>A booming economy</li>
  </ul>

  <h3 style="color: #333;">Levitt's Controversial Finding</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>The biggest factor was the legalization of abortion (<em>Roe v. Wade</em>, 1973)</li>
    <li>Unwanted children born in high-risk environments are more likely to become criminals</li>
    <li>18 years after 1973, those never-born children would have been peak crime age</li>
    <li>States that legalized abortion earlier saw crime drop earlier</li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> The real cause is often hidden — and controversial.</p>
  </blockquote>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Do Real Estate Agents Work for YOU?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Information Asymmetry Problem</em></p>

  <h3 style="color: #333;">The Problem</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Your agent earns ~1.5% commission on your home sale</li>
    <li>On a <span>$</span>300K house, an extra <span>$</span>10K for you = only <strong><span>$</span>150 more</strong> for them</li>
    <li>So their incentive is to sell <strong>fast</strong>, not sell <strong>high</strong></li>
  </ul>

  <h3 style="color: #333;">The Proof</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>When agents sell <strong>their own homes</strong>, they keep them on market <strong>10 extra days</strong></li>
    <li>And sell for <strong>3% more</strong> than comparable client homes!</li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> Incentives reveal true behavior.</p>
  </blockquote>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Teachers Cheating? Yes, Really!</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When High Stakes Create Perverse Incentives</em></p>

  <h3 style="color: #333;">Chicago Public Schools Standardized Testing</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>High-stakes testing: schools get rewarded or punished based on scores</li>
    <li>Teachers had access to answer sheets before submitting them</li>
  </ul>

  <h3 style="color: #333;">How Data Caught Them</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Levitt's algorithm found suspicious answer patterns: identical strings of correct answers, especially on hard questions</li>
    <li>Classrooms with dramatic score jumps followed by drops the next year</li>
    <li>Result: <strong>~5% of teachers were cheating</strong>. Many were fired!</li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> Algorithms can detect fraud humans miss.</p>
  </blockquote>

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Does Your Name Shape Your Destiny?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>A Tale of Two Brothers</em></p>

  <h3 style="color: #333;">The Story</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>A mother named her sons <strong>"Winner"</strong> and <strong>"Loser"</strong></li>
    <li>Winner became a criminal. Loser became a detective!</li>
  </ul>

  <h3 style="color: #333;">What the Data Actually Shows</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>Names don't <strong>cause</strong> success or failure</li>
    <li>But names <strong>signal</strong> the parents' education and economic status</li>
    <li>Names trickle down the economic ladder over time</li>
    <li>"Distinctly Black" names had <strong>50% lower</strong> callback rates on resumes (audit study) — revealing <strong>prejudice</strong>, not destiny</li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> Correlation is not causation!</p>
  </blockquote>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Do Drug Dealers Live with Their Moms?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When Real Data Destroys Hollywood Myths</em></p>

  <h3 style="color: #333;">The Myth</h3>
  <p style="font-size: 1.02em; color: #444;">Drug dealing is hugely profitable.</p>

  <h3 style="color: #333;">The Data (from actual gang financial records!)</h3>
  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li>A sociologist named Sudhir Venkatesh got access to a gang's books</li>
    <li>The gang was structured like McDonald's — a <strong>franchise model</strong>!</li>
    <li>Top boss: <strong><span>$</span>100K/year</strong>. Average foot soldier: <strong><span>$</span>3.30/hour</strong></li>
    <li>That's <strong>below minimum wage</strong> — hence living with Mom</li>
    <li><strong>1-in-4 chance</strong> of being killed in 4 years on the job</li>
  </ul>

  <blockquote style="border-left: 4px solid #4B2E83; padding: 10px 15px; margin: 15px 0; background-color: #F3EFF8;">
    <p style="font-style: italic; color: #555;"><strong>Data lesson:</strong> Real data destroys Hollywood myths.</p>
  </blockquote>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Art of Data Storytelling</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>What Makes Freakonomics a Storytelling Masterclass</em></p>

  <ul style="font-size: 1.02em; line-height: 1.9; color: #444;">
    <li><strong>Start with a question, not an answer</strong> — "Why do drug dealers live with their moms?" hooks attention instantly</li>
    <li><strong>Challenge assumptions with data</strong> — the sumo & teacher cheating stories show data overturning "common sense"</li>
    <li><strong>Use surprise as your weapon</strong> — unexpected findings (<span>$</span>3.30/hr for drug dealers) are unforgettable</li>
    <li><strong>Make data human</strong> — every chart tells a story about real people making real choices</li>
    <li><strong>Build a narrative arc</strong> — setup &rarr; data reveal &rarr; insight keeps audiences engaged from start to finish</li>
  </ul>

  <div style="background-color: #F3EFF8; border: 2px solid #4B2E83; padding: 15px 20px; border-radius: 8px; margin: 15px 0; text-align: center;">
    <p style="font-size: 1.15em; color: #4B2E83; font-weight: 600; margin: 0;">Great data storytelling = the right question + surprising data + a human narrative</p>
  </div>
