<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Tobler's First Law of Geography</h2>

> *"Everything is related to everything else, but near things are more related than distant things."*
> &mdash; Waldo Tobler, 1970

This deceptively simple statement is the **organizing principle** of spatial analysis. It tells us that:

- **Location matters.** The value of a variable at one place is related to the values at nearby places.
- **Observations are not independent.** Unlike a random sample of survey responses, spatial observations are connected through proximity, adjacency, and distance.
- **Patterns emerge from proximity.** Crime clusters in neighborhoods. House prices are similar on the same street. Pollution spreads from sources.

Spatial analysis provides tools to **detect, measure, and model** these spatial patterns.

**Key references:**
- Anselin, L. (1988). *Spatial Econometrics: Methods and Models*. Springer.
- LeSage, J. & Pace, R.K. (2009). *Introduction to Spatial Econometrics*. CRC Press.
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Spatial Data = Geometry + Attributes</h2>

A regular dataset:

| Neighborhood | Crime Rate | Income |
|:-------------|:-----------|:-------|
| A            | 45.2       | 32000  |
| B            | 12.8       | 58000  |

A **spatial** dataset adds a geometry column:

| Neighborhood | Crime Rate | Income | Geometry       |
|:-------------|:-----------|:-------|:---------------|
| A            | 45.2       | 32000  | POLYGON((...)) |
| B            | 12.8       | 58000  | POLYGON((...)) |

The geometry allows us to ask **spatial questions**:
- Where are the high values located?
- Are nearby places similar?
- Are there clusters? Outliers?
- Are events concentrated near roads, schools, or rivers?
- How do patterns change with scale?
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><strong>Problem:</strong> Do crime rates exhibit spatial clustering in Columbus? If nearby neighborhoods have similar crime rates, it suggests spatial autocorrelation &mdash; a direct manifestation of Tobler's First Law.</p>
  <p style="font-size: 0.95em; color: #555;"><strong>Data:</strong> Columbus Crime dataset &mdash; 49 neighborhoods, residential burglaries and vehicle thefts per 1000 households (1980). Source: Anselin (1988), <em>Spatial Econometrics</em>.</p>
  <p style="font-size: 0.95em; color: #555;"><strong>URL:</strong> <a href="https://geodacenter.github.io/data-and-lab/columbus/">geodacenter.github.io/data-and-lab/columbus/</a></p>
</div>

````python
# Example 1.1 — Columbus Crime: spatial pattern
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Choropleth of Crime
gdf_columbus.plot(ax=axes[0], column="CRIME", cmap="OrRd", edgecolor="black",
                  linewidth=0.5, legend=True,
                  legend_kwds={"label": "Crime Rate (per 1000 HH)", "shrink": 0.7})
axes[0].set_title("Columbus, OH: Crime Rate by Neighborhood\n(Anselin 1988)",
                  fontsize=12, fontweight="bold", color="#4B2E83")
axes[0].set_axis_off()

# Choropleth of Income
gdf_columbus.plot(ax=axes[1], column="INC", cmap="Greens", edgecolor="black",
                  linewidth=0.5, legend=True,
                  legend_kwds={"label": "Median Income ($1000s)", "shrink": 0.7})
axes[1].set_title("Columbus, OH: Median Income by Neighborhood",
                  fontsize=12, fontweight="bold", color="#4B2E83")
axes[1].set_axis_off()

plt.tight_layout()
plt.show()

print("Notice: high-crime areas (dark red) cluster together in the center.")
print("High-income areas (dark green) cluster on the periphery.")
print("Nearby neighborhoods tend to be similar — Tobler's First Law in action.")
print(f"\nCorrelation between CRIME and INC: {gdf_columbus['CRIME'].corr(gdf_columbus['INC']):.3f}")
````

````python
# Example 1.3 — Seattle: Histogram vs. Spatial Map of crime
fig = plt.figure(figsize=(18, 9))

# Bar chart — narrower column on the left
ax1 = fig.add_axes([0.02, 0.08, 0.35, 0.85])
top_cats = gdf_crime["offense_sub_category"].value_counts().head(8)
# Shorten long labels
short_labels = top_cats.index.str.replace(" (INCLUDES.*)", "", regex=True).str.title()
ax1.barh(range(len(top_cats)), top_cats.values, color="#4B2E83", edgecolor="white")
ax1.set_yticks(range(len(top_cats)))
ax1.set_yticklabels(short_labels, fontsize=11)
ax1.invert_yaxis()
ax1.set_xlabel("Count", fontsize=12)
ax1.set_title("Crime Counts by Category\n(no spatial information)",
              fontsize=13, fontweight="bold", color="#4B2E83")

# Map — wider column on the right with satellite basemap
ax2 = fig.add_axes([0.42, 0.05, 0.56, 0.90])
gdf_crime_web = gdf_crime.to_crs(epsg=3857)
gdf_crime_web.plot(ax=ax2, markersize=4, color="#F57C00", alpha=0.5, edgecolor="none")
ctx.add_basemap(ax2, source=ctx.providers.CartoDB.Voyager, zoom=12)
ax2.set_title("Crime Locations in Seattle\n(spatial clusters visible!)",
              fontsize=13, fontweight="bold", color="#4B2E83")
ax2.set_axis_off()

plt.show()

print("The bar chart tells us WHAT crimes occur most often.")
print("The map tells us WHERE crimes concentrate — downtown and specific corridors.")
print("Spatial analysis reveals patterns that tabular analysis cannot.")
````

| Type | Geometry | Real Examples in This Notebook | Common Analyses |
|:-----|:---------|:-------------------------------|:----------------|
| **Points** | (x, y) locations | Seattle crimes, Chicago carjackings | Distance, KDE, clustering, kriging |

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. What Is a Shapefile?</h2>

A **shapefile** is not a single file. It is a **group of files** that must stay together:

| File Extension | Purpose |
|:---------------|:--------|
| `.shp` | Stores the **geometry** (points, lines, or polygons) |
| `.shx` | Stores the **spatial index** (fast lookup of geometries) |
| `.dbf` | Stores the **attribute table** (like a spreadsheet) |
| `.prj` | Stores the **coordinate reference system** (projection info) |

**Optional supporting files:**

| File | Purpose |
|:-----|:--------|
| `.cpg` | Character encoding (e.g., UTF-8) |
| `.sbn` / `.sbx` | Spatial indexes (ESRI format) |
| `.xml` | Metadata |

**Important rule:** When sharing shapefiles, always **zip all related files together**.

**Other spatial formats:**

| Format | Extension | Advantages |
|:-------|:----------|:-----------|
| **GeoJSON** | `.geojson` | Human-readable, web-friendly, single file |
| **GeoPackage** | `.gpkg` | SQLite-based, single file, multiple layers |
| **GeoParquet** | `.parquet` | Fast columnar storage, cloud-friendly |
</div>

````python
# Example 3.1 — Inspect the Columbus shapefile components
import os

columbus_example = load_example("Columbus")
columbus_path = columbus_example.get_path("columbus.shp")
columbus_dir = os.path.dirname(columbus_path)

print("=== Columbus Shapefile Components ===")
print(f"Directory: {columbus_dir}\n")

columbus_files = sorted([f for f in os.listdir(columbus_dir) if "columbus" in f.lower()])
for f in columbus_files:
    ext = os.path.splitext(f)[1]
    size = os.path.getsize(os.path.join(columbus_dir, f))
    purpose = {".shp": "geometry", ".shx": "spatial index", ".dbf": "attributes",
               ".prj": "CRS/projection"}.get(ext, "other")
    print(f"  {f:25s} {size:>8,} bytes  ({purpose})")

print(f"\nReading with GeoPandas:")
gdf = gpd.read_file(columbus_path)
print(f"  Rows: {len(gdf)}, Columns: {len(gdf.columns)}")
print(f"  CRS: {gdf.crs}")
print(f"  Geometry type: {gdf.geom_type.unique()}")
````

````python
# Example 3.2 — Compare file formats: save Columbus in different formats
import tempfile

tmp = tempfile.mkdtemp()

# Save in 3 formats
gdf_columbus.to_file(os.path.join(tmp, "columbus.shp"))
gdf_columbus.to_file(os.path.join(tmp, "columbus.geojson"), driver="GeoJSON")
gdf_columbus.to_file(os.path.join(tmp, "columbus.gpkg"), driver="GPKG")

# Compare
print("=== File Format Comparison (Columbus, 49 neighborhoods) ===\n")
print(f"{'Format':<12} {'Files':<8} {'Total Size':>12}")
print("-" * 35)

# Shapefile
shp_files = [f for f in os.listdir(tmp) if f.startswith("columbus") and not f.endswith((".geojson", ".gpkg"))]
shp_size = sum(os.path.getsize(os.path.join(tmp, f)) for f in shp_files)
print(f"{'Shapefile':<12} {len(shp_files):<8} {shp_size:>10,} B")

# GeoJSON
gj_size = os.path.getsize(os.path.join(tmp, "columbus.geojson"))
print(f"{'GeoJSON':<12} {'1':<8} {gj_size:>10,} B")

# GeoPackage
gpkg_size = os.path.getsize(os.path.join(tmp, "columbus.gpkg"))
print(f"{'GeoPackage':<12} {'1':<8} {gpkg_size:>10,} B")

print("\nShapefile = multiple files (must zip together)")
print("GeoJSON   = single file, human-readable, good for web")
print("GeoPackage = single file, efficient, recommended for new projects")
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Geographic vs. Projected CRS</h2>

| Aspect | Geographic CRS | Projected CRS |
|:-------|:---------------|:--------------|
| **Units** | Degrees (latitude/longitude) | Meters or feet |
| **Shape** | Coordinates on a sphere/ellipsoid | Coordinates on a flat surface |
| **Example** | EPSG:4326 (WGS84) | EPSG:32610 (UTM Zone 10N) |
| **Use** | Storage, GPS data, web APIs | Distance, area, buffers, density |

**Critical rule:** Never calculate distance or area directly in degrees (lat/lon). Always project first.

**Common EPSG codes:**
- **EPSG:4326** &mdash; WGS84, used by GPS and most web data
- **EPSG:3857** &mdash; Web Mercator (Google Maps, OSM tiles)
- **EPSG:32610** &mdash; UTM Zone 10N (Seattle/Pacific NW, meters)
- **EPSG:5070** &mdash; NAD83 Conus Albers (US equal-area)

**Projection types and what they preserve:**

| Type | Preserves | Distorts | Use Case |
|:-----|:----------|:---------|:---------|
| **Equal-area** (Albers, Mollweide) | Area | Shape | Choropleth maps, area calculations |
| **Conformal** (Mercator, Lambert) | Shape, angles | Area | Navigation, web maps |
| **Equidistant** (Azimuthal) | Distance from center | Area, shape | Measuring distance |
| **UTM** | Local shape + area | Nothing locally | Engineering, local analysis |
</div>

| Method | Description | Best For |
|:-------|:-----------|:---------|
| **Equal Interval** | Equal-width bins | Uniform distributions |
| **Quantiles** | Equal count per bin | Skewed data |
| **Natural Breaks (Jenks)** | Minimizes within-class variance | Finding natural groupings |
| **Standard Deviation** | Bins around the mean | Showing deviation from average |

**Color palette guidelines:**
- **Sequential** (light→dark): Ordered data (crime rate, population)
- **Diverging** (two extremes): Deviations from center (above/below average)
- **Categorical** (distinct): Nominal classes (land use types)
- **Avoid rainbow colormaps** — not perceptually uniform

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Types of Spatial Weights</h2>

| Type | Definition | Best For |
|:-----|:-----------|:---------|
| **Queen contiguity** | Share an edge OR vertex | Polygons (most common) |
| **Rook contiguity** | Share an edge only | Regular grids |
| **K-Nearest Neighbors** | k closest centroids | Points or irregular polygons |
| **Distance band** | Within a threshold distance | Points, varying density |

**Row-standardization:** Divide each row by its sum so weights sum to 1. The spatial lag then represents the **average** of neighbor values.
</div>

````python
# Example 7.1 — Queen vs. Rook contiguity (Columbus)
w_queen = Queen.from_dataframe(gdf_columbus)
w_rook = Rook.from_dataframe(gdf_columbus)

print("=== Spatial Weights: Columbus ===")
print(f"\nQueen: avg neighbors = {w_queen.mean_neighbors:.2f}, "
      f"min = {w_queen.min_neighbors}, max = {w_queen.max_neighbors}")
print(f"Rook:  avg neighbors = {w_rook.mean_neighbors:.2f}, "
      f"min = {w_rook.min_neighbors}, max = {w_rook.max_neighbors}")

# Histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
queen_card = [len(w_queen[i]) for i in range(w_queen.n)]
rook_card = [len(w_rook[i]) for i in range(w_rook.n)]

axes[0].hist(queen_card, bins=range(1, 12), color="#4B2E83", edgecolor="white", align="left")
axes[0].set_title("Queen Contiguity (edge + vertex)", fontsize=11, fontweight="bold", color="#4B2E83")
axes[0].set_xlabel("Number of Neighbors")
axes[0].set_ylabel("Count")

axes[1].hist(rook_card, bins=range(1, 12), color="#E8B923", edgecolor="white", align="left")
axes[1].set_title("Rook Contiguity (edge only)", fontsize=11, fontweight="bold", color="#4B2E83")
axes[1].set_xlabel("Number of Neighbors")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
````

````python
# Example 7.2 — Visualizing Queen connectivity network (Columbus)
w = Queen.from_dataframe(gdf_columbus)

fig, ax = plt.subplots(figsize=(9, 8))
gdf_columbus.plot(ax=ax, color="#F9F7FC", edgecolor="#4B2E83", linewidth=0.8)

centroids = gdf_columbus.geometry.centroid
for i in range(w.n):
    for j in w[i]:
        ax.plot([centroids.iloc[i].x, centroids.iloc[j].x],
                [centroids.iloc[i].y, centroids.iloc[j].y],
                color="#F57C00", linewidth=0.8, alpha=0.5)

ax.scatter(centroids.x, centroids.y, s=20, c="#4B2E83", zorder=5)
ax.set_title("Queen Contiguity Network — Columbus\n(orange = neighbor connections)",
             fontsize=13, fontweight="bold", color="#4B2E83")
ax.set_axis_off()
plt.tight_layout()
plt.show()

print("Each polygon is connected to its Queen neighbors.")
print("This structure defines 'nearby' for Moran's I and LISA.")
````

  <p style="color: #E8D3FF; font-family: 'Encode Sans', 'Arial', sans-serif; font-size: 1.3em; margin: 10px 0; font-weight: 300;">Moran's I &mdash; Are Similar Values Near Each Other?</p>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Interpreting Moran's I</h2>

| Moran's I Value | Meaning | Pattern |
|:----------------|:--------|:--------|
| **I > 0** | Positive spatial autocorrelation | Similar values cluster (HH, LL) |
| **I < 0** | Negative spatial autocorrelation | Dissimilar values are neighbors (HL, LH) |
| **I &approx; 0** | No spatial autocorrelation | Spatial randomness |

**Connection to Tobler's Law:** If Moran's I is significantly positive, nearby places are indeed more similar than distant places.
</div>

````python
# Example 8.1 — Global Moran's I for Columbus Crime
w = Queen.from_dataframe(gdf_columbus)
w.transform = "R"

moran_crime = Moran(gdf_columbus["CRIME"].values, w)

print("=== Global Moran's I: Columbus Crime ===")
print(f"Moran's I:      {moran_crime.I:.4f}")
print(f"Expected I:     {moran_crime.EI:.4f}")
print(f"p-value (sim):  {moran_crime.p_sim:.4f}")
print(f"z-score:        {moran_crime.z_sim:.4f}")
print(f"Permutations:   {moran_crime.permutations}")
print()
if moran_crime.p_sim < 0.05 and moran_crime.I > 0:
    print("CONCLUSION: Significant POSITIVE spatial autocorrelation.")
    print("Crime is spatially clustered — Tobler's Law confirmed for this variable.")
````

````python
# Example 8.2 — Moran Scatterplot
fig, ax = plt.subplots(figsize=(8, 7))
moran_scatterplot(moran_crime, ax=ax, aspect_equal=False)
ax.set_title(f"Moran Scatterplot: Columbus Crime\n(I = {moran_crime.I:.4f}, p = {moran_crime.p_sim:.4f})",
             fontsize=13, fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

print("Quadrants:")
print("  Upper-Right (HH): High crime, high-crime neighbors")
print("  Lower-Left  (LL): Low crime, low-crime neighbors")
print("  Upper-Left  (LH): Low crime, high-crime neighbors (spatial outlier)")
print("  Lower-Right (HL): High crime, low-crime neighbors (spatial outlier)")
````

````python
# Example 8.3 — Compare Moran's I across variables and datasets
# Columbus
w_col = Queen.from_dataframe(gdf_columbus)
w_col.transform = "R"
results = []
for var in ["CRIME", "INC", "HOVAL"]:
    m = Moran(gdf_columbus[var].values, w_col)
    results.append({"Dataset": "Columbus", "Variable": var, "Moran_I": m.I, "p_value": m.p_sim})

# Guerry
w_gur = Queen.from_dataframe(gdf_guerry)
w_gur.transform = "R"
for var in ["Crm_prs", "Litercy", "Suicids"]:
    m = Moran(gdf_guerry[var].values, w_gur)
    results.append({"Dataset": "Guerry", "Variable": var, "Moran_I": m.I, "p_value": m.p_sim})

results_df = pd.DataFrame(results)
print("=== Moran's I Comparison Across Real Datasets ===\n")
print(results_df.to_string(index=False))
print("\nAll variables show significant positive spatial autocorrelation.")
print("Tobler's Law holds across different places and time periods!")
````

| Cluster Type | Meaning | Color |
|:-------------|:--------|:------|
| **High-High (HH)** | High value, high-value neighbors | Red |
| **Low-Low (LL)** | Low value, low-value neighbors | Blue |
| **High-Low (HL)** | High value, low-value neighbors (outlier) | Pink |
| **Low-High (LH)** | Low value, high-value neighbors (outlier) | Light blue |
| **Not Significant** | No significant local pattern | Gray |

**Applications:** Crime hotspots (HH), poverty traps (LL), gentrifying areas (LH/HL).
</div>

````python
# Example 9.1 — LISA for Columbus Crime
w = Queen.from_dataframe(gdf_columbus)
w.transform = "R"
lisa = Moran_Local(gdf_columbus["CRIME"].values, w)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Choropleth
gdf_columbus.plot(ax=axes[0], column="CRIME", cmap="OrRd", edgecolor="black",
                  linewidth=0.5, legend=True)
axes[0].set_title("Choropleth: Crime Rate", fontsize=11, fontweight="bold", color="#4B2E83")
axes[0].set_axis_off()

# LISA cluster map
lisa_cluster(lisa, gdf_columbus, p=0.05, ax=axes[1])
axes[1].set_title("LISA Clusters: Crime (p < 0.05)", fontsize=11, fontweight="bold", color="#4B2E83")
axes[1].set_axis_off()

plt.suptitle("Columbus Crime — Choropleth vs. LISA",
             fontsize=13, fontweight="bold", color="#4B2E83", y=1.01)
plt.tight_layout()
plt.show()

sig = lisa.p_sim < 0.05
labels = lisa.q
cluster_names = {1: "HH (Hot Spot)", 2: "LH (Outlier)", 3: "LL (Cold Spot)", 4: "HL (Outlier)"}
print("\n=== Significant LISA Clusters (p < 0.05) ===")
for q in [1, 2, 3, 4]:
    count = ((labels == q) & sig).sum()
    if count > 0:
        print(f"  {cluster_names[q]}: {count} neighborhoods")
print(f"  Not Significant: {(~sig).sum()} neighborhoods")
````

````python
# Example 10.1 — Nearest Neighbor Analysis on Chicago Carjackings
coords = np.column_stack([gdf_carjackings.geometry.x, gdf_carjackings.geometry.y])

nn = NearestNeighbors(n_neighbors=2)
nn.fit(coords)
distances, indices = nn.kneighbors(coords)
nn_distances = distances[:, 1]  # distance to nearest neighbor

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

gdf_cj = gdf_carjackings.copy()
gdf_cj["nn_dist"] = nn_distances
gdf_cj.plot(ax=axes[0], column="nn_dist", cmap="viridis_r", markersize=8,
            edgecolor="none", legend=True,
            legend_kwds={"label": "NN Distance (degrees)", "shrink": 0.7})
axes[0].set_title("Chicago Carjackings\nColored by Nearest Neighbor Distance",
                  fontsize=11, fontweight="bold", color="#4B2E83")

axes[1].hist(nn_distances, bins=40, color="#4B2E83", edgecolor="white", alpha=0.8)
axes[1].axvline(nn_distances.mean(), color="#F57C00", linewidth=2, linestyle="--",
                label=f"Mean = {nn_distances.mean():.5f}°")
axes[1].set_title("Distribution of NN Distances", fontsize=11, fontweight="bold", color="#4B2E83")
axes[1].set_xlabel("Distance to Nearest Neighbor")
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"Mean NN distance: {nn_distances.mean():.6f}°")
print(f"Median NN distance: {np.median(nn_distances):.6f}°")
print("Short NN distances indicate spatial clustering of carjackings.")
````

````python
# Example 10.2 — DBSCAN Clustering on Chicago Carjackings
coords = np.column_stack([gdf_carjackings.geometry.x, gdf_carjackings.geometry.y])

# DBSCAN with eps in degrees (≈0.01° ≈ 1.1 km)
db = DBSCAN(eps=0.01, min_samples=10)
labels = db.fit_predict(coords)
n_clusters = len(set(labels) - {-1})

gdf_cj = gdf_carjackings.copy()
gdf_cj["cluster"] = labels

fig, ax = plt.subplots(figsize=(8, 10))

# Noise in gray
noise = gdf_cj[gdf_cj["cluster"] == -1]
clustered = gdf_cj[gdf_cj["cluster"] >= 0]

noise.plot(ax=ax, markersize=3, color="#CCCCCC", alpha=0.3, label="Noise")
clustered.plot(ax=ax, column="cluster", cmap="tab10", markersize=8,
               edgecolor="none", alpha=0.6, legend=False)

ax.set_title(f"DBSCAN: Chicago Carjackings\n{n_clusters} clusters, "
             f"{len(noise)} noise points",
             fontsize=13, fontweight="bold", color="#4B2E83")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.tight_layout()
plt.show()

print(f"DBSCAN found {n_clusters} carjacking hotspot clusters.")
print(f"Points in clusters: {len(clustered)} ({100*len(clustered)/len(gdf_cj):.0f}%)")
print(f"Noise points: {len(noise)} ({100*len(noise)/len(gdf_cj):.0f}%)")
````

````python
# Example 10.3 — Kriging: Interpolating Columbus Crime
from pykrige.ok import OrdinaryKriging

# Use Columbus centroids and crime values
centroids = gdf_columbus.geometry.centroid
x_obs = centroids.x.values
y_obs = centroids.y.values
z_obs = gdf_columbus["CRIME"].values

# Kriging
OK = OrdinaryKriging(x_obs, y_obs, z_obs, variogram_model="spherical",
                     verbose=False, enable_plotting=False)

gridx = np.linspace(x_obs.min(), x_obs.max(), 50)
gridy = np.linspace(y_obs.min(), y_obs.max(), 50)
z_pred, ss_pred = OK.execute("grid", gridx, gridy)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

im1 = axes[0].contourf(gridx, gridy, z_pred, levels=15, cmap="OrRd")
axes[0].scatter(x_obs, y_obs, c="black", s=15, edgecolor="white", linewidth=0.3, zorder=5)
plt.colorbar(im1, ax=axes[0], label="Predicted Crime Rate", shrink=0.7)
axes[0].set_title("Kriging: Predicted Crime Surface\n(Columbus, interpolated from 49 neighborhoods)",
                  fontsize=10, fontweight="bold", color="#4B2E83")

im2 = axes[1].contourf(gridx, gridy, np.sqrt(ss_pred), levels=15, cmap="Purples")
axes[1].scatter(x_obs, y_obs, c="black", s=15, edgecolor="white", linewidth=0.3, zorder=5)
plt.colorbar(im2, ax=axes[1], label="Uncertainty (std)", shrink=0.7)
axes[1].set_title("Kriging: Prediction Uncertainty\n(higher where data is sparse)",
                  fontsize=10, fontweight="bold", color="#4B2E83")

plt.tight_layout()
plt.show()

print("Kriging uses Tobler's Law: nearby values predict unknown locations.")
print("Uncertainty increases where observations are sparse.")
````

  <h1 style="color: #FFF; font-family: 'Encode Sans', 'Arial', sans-serif; font-weight: 700; margin: 0; text-shadow: 2px 2px 4px rgba(0,0,0,0.3);">Part 13: Common Mistakes in Spatial Visualization</h1>
  <p style="color: #E8D3FF; font-family: 'Encode Sans', 'Arial', sans-serif; font-size: 1.3em; margin: 10px 0; font-weight: 300;">Projection Errors, MAUP, and Ecological Fallacy</p>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Common Mistakes</h2>

| Mistake | Problem | Solution |
|:--------|:--------|:---------|
| Calculating distance in degrees | 1° ≠ constant distance | Always project first |
| Using Mercator for area | Extreme distortion at poles | Use equal-area projection |
| Mapping raw counts | Large areas/populations dominate | Normalize by population or area |
| Rainbow colormap | False boundaries, not uniform | Use sequential/diverging |
| **MAUP** | Different boundaries → different results | Test at multiple scales |
| **Ecological fallacy** | Area patterns ≠ individual behavior | Never infer individual from aggregate |
</div>

````python
# Example 13.1 — MAUP with Columbus Crime
# Aggregate Columbus neighborhoods into different groupings
gdf = gdf_columbus.copy()

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Original 49 neighborhoods
gdf.plot(ax=axes[0], column="CRIME", cmap="OrRd", edgecolor="black", linewidth=0.3, legend=True)
axes[0].set_title(f"49 Neighborhoods\nmean crime = {gdf['CRIME'].mean():.1f}",
                  fontsize=10, fontweight="bold", color="#4B2E83")
axes[0].set_axis_off()

# Group into ~12 regions (by POLYID quartiles)
gdf["region4"] = pd.qcut(gdf["POLYID"], 4, labels=["NW", "NE", "SW", "SE"])
dissolved4 = gdf.dissolve(by="region4", aggfunc="mean").reset_index()
dissolved4.plot(ax=axes[1], column="CRIME", cmap="OrRd", edgecolor="black", linewidth=1, legend=True)
axes[1].set_title(f"4 Regions (dissolved)\nmean crime = {dissolved4['CRIME'].mean():.1f}",
                  fontsize=10, fontweight="bold", color="#4B2E83")
axes[1].set_axis_off()

# Group into 2 halves
gdf["half"] = gdf["POLYID"].apply(lambda x: "West" if x <= 25 else "East")
dissolved2 = gdf.dissolve(by="half", aggfunc="mean").reset_index()
dissolved2.plot(ax=axes[2], column="CRIME", cmap="OrRd", edgecolor="black", linewidth=1.5, legend=True)
axes[2].set_title(f"2 Halves (dissolved)\nmean crime = {dissolved2['CRIME'].mean():.1f}",
                  fontsize=10, fontweight="bold", color="#4B2E83")
axes[2].set_axis_off()

plt.suptitle("MAUP: Same Crime Data at Different Aggregation Levels",
             fontsize=13, fontweight="bold", color="#4B2E83", y=1.02)
plt.tight_layout()
plt.show()

print("MAUP: the spatial pattern CHANGES when you change the zone boundaries.")
print("More zones → more spatial detail. Fewer zones → smoother but lose local patterns.")
````

````python
# Example 13.2 — Raw counts vs. rates (Columbus)
# Simulate: what if we only had raw crime COUNTS (not rates)?
gdf = gdf_columbus.copy()
gdf_proj = gdf.to_crs(gdf.estimate_utm_crs())
gdf_proj["area_km2"] = gdf_proj.geometry.area / 1_000_000

# Pretend: raw_crimes = crime_rate * area (larger areas accumulate more)
gdf["raw_crimes"] = gdf["CRIME"] * gdf_proj["area_km2"].values

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

gdf.plot(ax=axes[0], column="raw_crimes", cmap="OrRd", edgecolor="black",
         linewidth=0.5, legend=True, legend_kwds={"shrink": 0.7})
axes[0].set_title("MISLEADING: Raw Crime Counts\n(larger neighborhoods dominate)",
                  fontsize=10, fontweight="bold", color="red")
axes[0].set_axis_off()

gdf.plot(ax=axes[1], column="CRIME", cmap="OrRd", edgecolor="black",
         linewidth=0.5, legend=True, legend_kwds={"shrink": 0.7})
axes[1].set_title("CORRECT: Crime Rate (per 1000 HH)\n(normalized — comparable across areas)",
                  fontsize=10, fontweight="bold", color="#2E8B57")
axes[1].set_axis_off()

plt.tight_layout()
plt.show()

print("Always normalize: rates per capita, per area, or per exposure unit.")
print("Raw counts make large areas/populations look worse regardless of actual risk.")
````

````python
# Example 13.3 — Ecological Fallacy (using Columbus data)
print("=== ECOLOGICAL FALLACY ===")
print()
print("Columbus data shows (at NEIGHBORHOOD level):")
r = gdf_columbus["CRIME"].corr(gdf_columbus["INC"])
print(f"  Correlation(Crime, Income) = {r:.3f}")
print(f"  Higher income neighborhoods → lower crime rates")
print()
print("WRONG INFERENCE (ecological fallacy):")
print("  'High-income individuals commit less crime'")
print()
print("WHY THIS IS WRONG:")
print("  - The data is about NEIGHBORHOODS, not individuals")
print("  - Low-income areas may have crime due to: lack of policing,")
print("    lack of street lighting, vacant properties, opportunity structure")
print("  - Many low-income individuals in high-crime areas never commit crimes")
print("  - Many high-income individuals commit white-collar crimes")
print()
print("CORRECT STATEMENT:")
print("  'Neighborhoods with lower median income tend to have higher crime rates.'")
print("  'This area-level pattern cannot be attributed to individual behavior.'")

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(gdf_columbus["INC"], gdf_columbus["CRIME"], s=60, c="#4B2E83",
           edgecolor="black", linewidth=0.5)
z = np.polyfit(gdf_columbus["INC"], gdf_columbus["CRIME"], 1)
x_line = np.linspace(gdf_columbus["INC"].min(), gdf_columbus["INC"].max(), 100)
ax.plot(x_line, np.poly1d(z)(x_line), color="#F57C00", linewidth=2, linestyle="--")
ax.set_xlabel("Neighborhood Median Income ($1000s)")
ax.set_ylabel("Neighborhood Crime Rate (per 1000 HH)")
ax.set_title(f"Area-Level Correlation (r = {r:.3f})\nDoes NOT prove individual-level relationship!",
             fontsize=12, fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()
````
