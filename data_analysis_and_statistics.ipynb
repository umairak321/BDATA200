  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">🎲 Sample Space, Sigma-Algebra, Events & Probability</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Probability Triple (Ω, 𝓕, P)</em></p>
</div>

### **SAMPLE SPACE (Ω)**
- The set of **all possible outcomes** of a random experiment
- Coin toss: Ω = {H, T}
- Die roll: Ω = {1, 2, 3, 4, 5, 6}
- Two coins: Ω = {HH, HT, TH, TT}

### **EVENTS**
- An **event** is a subset of the sample space (A ⊆ Ω)
- "Rolling an even number": A = {2, 4, 6}
- The empty set ∅ and Ω itself are always events

### **SIGMA-ALGEBRA (𝓕)**
- A collection of events that we agree to measure
- Must contain Ω, be closed under complements and countable unions
- Think of it as the **"rule book"** of what questions we can ask

### **PROBABILITY FUNCTION P**
- P: 𝓕 → [0, 1] assigns a likelihood to each event
- P(Ω) = 1 (something must happen)
- P(∅) = 0 (impossible event)
- For disjoint events: P(A ∪ B) = P(A) + P(B)

  <p style="margin: 10px 0 0 0; color: #333;">The Probability Triple (Ω, 𝓕, P) — The sample space, the events we care about, and how likely each is.</p>

```python
# --- Example 1: Simulating a fair coin ---
n_flips = 100_000
flips = np.random.choice(['H', 'T'], size=n_flips)

p_heads = np.mean(flips == 'H')
print(f'Sample space Ω = {{H, T}}')
print(f'P(Heads) after {n_flips:,} flips = {p_heads:.4f}  (theory: 0.5)')
```

```python
# --- Example 2: Simulating a fair die ---
n_rolls = 100_000
rolls = np.random.randint(1, 7, size=n_rolls)

# Sample space
omega = {1, 2, 3, 4, 5, 6}
print(f'Sample space Ω = {omega}')

# Event A: rolling an even number
event_A = {2, 4, 6}
p_even = np.mean(np.isin(rolls, list(event_A)))
print(f'Event A (even) = {event_A}')
print(f'P(even) = {p_even:.4f}  (theory: 0.5)')

# Event B: rolling greater than 4
event_B = {5, 6}
p_gt4 = np.mean(np.isin(rolls, list(event_B)))
print(f'Event B (>4) = {event_B}')
print(f'P(>4) = {p_gt4:.4f}  (theory: {len(event_B)/len(omega):.4f})')
```

```python
# --- Example 3: Sigma-algebra and event relationships ---
# For a die, the power set of Ω has 2^6 = 64 elements
# But we'll illustrate with a simpler case: coin toss

omega_coin = {'H', 'T'}
# The sigma-algebra for a coin (the power set):
sigma_algebra = [set(), {'H'}, {'T'}, {'H', 'T'}]
print('Sigma-algebra 𝓕 for a fair coin:')
for event in sigma_algebra:
    label = '∅' if len(event) == 0 else str(event)
    print(f'  {label}')

# Verify closure under complement
A = {'H'}
A_complement = omega_coin - A
print(f'\nA = {A},  Complement of A = {A_complement}')
print(f'Both in 𝓕? {A in sigma_algebra and A_complement in sigma_algebra}')
```

```python
# --- Example 4: Two dice — larger sample space ---
from itertools import product

omega_2dice = list(product(range(1,7), range(1,7)))
print(f'Sample space for 2 dice: |Ω| = {len(omega_2dice)}')
print(f'First 6 outcomes: {omega_2dice[:6]}')

# Simulate sum of two dice
n_sim = 100_000
die1 = np.random.randint(1, 7, n_sim)
die2 = np.random.randint(1, 7, n_sim)
sums = die1 + die2

# P(sum = 7)
p_sum7 = np.mean(sums == 7)
theory_p7 = 6 / 36  # 6 ways to make 7
print(f'\nP(sum=7) simulated = {p_sum7:.4f}, theory = {theory_p7:.4f}')

# Plot distribution of sums
fig, ax = plt.subplots(figsize=(10, 5))
values, counts = np.unique(sums, return_counts=True)
ax.bar(values, counts / n_sim, color='#4B2E83', edgecolor='white', alpha=0.85)
ax.set_xlabel('Sum of Two Dice')
ax.set_ylabel('Probability')
ax.set_title('Probability Distribution of the Sum of Two Dice', fontsize=14, color='#4B2E83')
ax.set_xticks(range(2, 13))
plt.tight_layout()
plt.show()
```

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">📈 Discrete Random Variables</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Countable Outcomes Described by a PMF</em></p>
</div>

### **What is a Random Variable?**
- A function X: Ω → ℝ that maps outcomes to numbers
- It lets us **do math** with randomness

### **Discrete Random Variables**
- Takes **countable** values (finite or infinite)
- Described by a **PMF** (Probability Mass Function): P(X = x)
- Examples: number of emails per day, coin flips (0 or 1), defects in a batch

### **Common Discrete Distributions**
- **Bernoulli**: single yes/no trial
- **Binomial**: number of successes in n trials
- **Poisson**: count of events in a fixed interval
- **Geometric**: number of trials until first success

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">P(X = x) gives the exact probability of each value for discrete random variables.</p>
</div>

```python
# --- Discrete distributions: PMF plots ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Bernoulli
x_bern = [0, 1]
axes[0,0].bar(x_bern, [0.4, 0.6], color='#4B2E83', edgecolor='white')
axes[0,0].set_title('Bernoulli (p=0.6)', fontsize=13, color='#4B2E83')
axes[0,0].set_xticks([0, 1])
axes[0,0].set_ylabel('P(X = x)')

# Binomial
k = np.arange(0, 21)
axes[0,1].bar(k, stats.binom.pmf(k, n=20, p=0.5), color='#E8B923', edgecolor='white')
axes[0,1].set_title('Binomial (n=20, p=0.5)', fontsize=13, color='#4B2E83')
axes[0,1].set_ylabel('P(X = x)')

# Poisson
k2 = np.arange(0, 20)
axes[1,0].bar(k2, stats.poisson.pmf(k2, mu=5), color='#32006E', edgecolor='white')
axes[1,0].set_title('Poisson (λ=5)', fontsize=13, color='#4B2E83')
axes[1,0].set_ylabel('P(X = x)')

# Geometric
k3 = np.arange(1, 16)
axes[1,1].bar(k3, stats.geom.pmf(k3, p=0.3), color='#85714D', edgecolor='white')
axes[1,1].set_title('Geometric (p=0.3)', fontsize=13, color='#4B2E83')
axes[1,1].set_ylabel('P(X = x)')

for ax in axes.flat:
    ax.set_xlabel('x')
plt.suptitle('Common Discrete Distributions (PMF)', fontsize=16, color='#4B2E83', y=1.02)
plt.tight_layout()
plt.show()
```

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">📊 Continuous Random Variables</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Uncountable Outcomes Described by a PDF</em></p>
</div>

### **Continuous Random Variables**
- Takes **uncountable** values (any real number in a range)
- Described by a **PDF** (Probability Density Function)
- P(X = x) = 0 for any single point!
- Use P(a ≤ X ≤ b) = area under the curve

### **Common Continuous Distributions**
- **Normal (Gaussian)**: bell-shaped, symmetric
- **Exponential**: time between events
- **Uniform**: equal probability across a range
- **Log-normal**: right-skewed, common in finance

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">For continuous random variables, probability = area under the PDF curve. Use the CDF for cumulative probabilities.</p>
</div>

```python
# --- Continuous distributions: PDF plots ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
x = np.linspace(-4, 4, 300)

# Normal
axes[0,0].plot(x, stats.norm.pdf(x, 0, 1), color='#4B2E83', lw=2.5)
axes[0,0].fill_between(x, stats.norm.pdf(x, 0, 1), alpha=0.3, color='#4B2E83')
axes[0,0].set_title('Normal (μ=0, σ=1)', fontsize=13, color='#4B2E83')

# Exponential
x2 = np.linspace(0, 5, 300)
axes[0,1].plot(x2, stats.expon.pdf(x2, scale=1), color='#E8B923', lw=2.5)
axes[0,1].fill_between(x2, stats.expon.pdf(x2, scale=1), alpha=0.3, color='#E8B923')
axes[0,1].set_title('Exponential (λ=1)', fontsize=13, color='#4B2E83')

# Uniform
x3 = np.linspace(-0.5, 1.5, 300)
axes[1,0].plot(x3, stats.uniform.pdf(x3, 0, 1), color='#32006E', lw=2.5)
axes[1,0].fill_between(x3, stats.uniform.pdf(x3, 0, 1), alpha=0.3, color='#32006E')
axes[1,0].set_title('Uniform (0, 1)', fontsize=13, color='#4B2E83')

# Log-Normal
x4 = np.linspace(0.01, 8, 300)
axes[1,1].plot(x4, stats.lognorm.pdf(x4, s=0.5), color='#85714D', lw=2.5)
axes[1,1].fill_between(x4, stats.lognorm.pdf(x4, s=0.5), alpha=0.3, color='#85714D')
axes[1,1].set_title('Log-Normal (σ=0.5)', fontsize=13, color='#4B2E83')

for ax in axes.flat:
    ax.set_ylabel('f(x)')
    ax.set_xlabel('x')
plt.suptitle('Common Continuous Distributions (PDF)', fontsize=16, color='#4B2E83', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# --- Applied example: Histogram + CDF ---
# Simulate daily temperatures (Normal distribution)
np.random.seed(42)
temperatures = np.random.normal(loc=72, scale=8, size=1000)  # °F

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(temperatures, bins=30, color='#4B2E83', edgecolor='white',
             alpha=0.8, density=True)
xfit = np.linspace(40, 100, 200)
axes[0].plot(xfit, stats.norm.pdf(xfit, 72, 8), 'r-', lw=2, label='Normal fit')
axes[0].set_title('Daily Temperature Distribution', fontsize=14, color='#4B2E83')
axes[0].set_xlabel('Temperature (°F)')
axes[0].set_ylabel('Density')
axes[0].legend()

# CDF
sorted_temps = np.sort(temperatures)
cdf = np.arange(1, len(sorted_temps)+1) / len(sorted_temps)
axes[1].plot(sorted_temps, cdf, color='#E8B923', lw=2)
axes[1].plot(xfit, stats.norm.cdf(xfit, 72, 8), 'r--', lw=2, label='Theoretical CDF')
axes[1].set_title('Cumulative Distribution Function', fontsize=14, color='#4B2E83')
axes[1].set_xlabel('Temperature (°F)')
axes[1].set_ylabel('Cumulative Probability')
axes[1].axhline(0.5, color='gray', ls=':', alpha=0.5)
axes[1].legend()

plt.tight_layout()
plt.show()

# Probability calculations
print(f'P(Temp > 80°F) = {np.mean(temperatures > 80):.4f}')
print(f'P(60 < Temp < 84) = {np.mean((temperatures > 60) & (temperatures < 84)):.4f}')
```

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">📊 Population vs Sample</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Science of Learning from Data</em></p>
</div>

### **POPULATION vs SAMPLE**
- **Population**: the entire group of interest (often too large to measure)
- **Sample**: a subset we actually observe
- Why sample? Cost, time, feasibility

### **PARAMETER vs STATISTIC**
- **Parameter**: a fixed (unknown) number about the population (μ, σ)
- **Statistic**: a number computed from the sample (x̄, s)
- Statistics **estimate** parameters
- μ (population mean) vs x̄ (sample mean)
- σ (population std dev) vs s (sample std dev)

### **TYPES OF DATA**
- **Quantitative**: Continuous (height, income) or Discrete (count of items)
- **Qualitative**: Nominal (color, city) or Ordinal (rating: 1–5)

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Statistics uses sample data to draw conclusions about populations under uncertainty.</p>
</div>

```python
# --- Qualitative and Quantitative data in pandas ---
np.random.seed(42)

# Create a sample dataset
n = 200
df = pd.DataFrame({
    # Quantitative - Continuous
    'height_cm': np.random.normal(170, 10, n).round(1),
    'income': np.random.lognormal(10.5, 0.5, n).round(2),
    # Quantitative - Discrete
    'num_children': np.random.poisson(1.5, n),
    # Qualitative - Nominal
    'city': np.random.choice(['Seattle', 'Portland', 'San Francisco', 'Denver'], n),
    # Qualitative - Ordinal
    'satisfaction': np.random.choice(['Very Low', 'Low', 'Medium', 'High', 'Very High'], n)
})

# Set ordinal category
df['satisfaction'] = pd.Categorical(
    df['satisfaction'],
    categories=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
    ordered=True
)

print('--- Data Types ---')
print(df.dtypes)
print('\n--- First 5 rows ---')
df.head()
```

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">🎯 Sampling Distribution of the Mean</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Proving the Sample Mean is Unbiased</em></p>
</div>

```python
# --- Population of 10 elements, all samples of size 3 ---
from itertools import combinations

population = np.array([3, 7, 11, 15, 18, 21, 25, 30, 34, 40])
pop_mean = population.mean()
print(f'Population: {population}')
print(f'Population mean μ = {pop_mean}')
print(f'Population size N = {len(population)}')

# Generate ALL samples of size 3
sample_size = 3
all_samples = list(combinations(population, sample_size))
n_samples = len(all_samples)
print(f'\nNumber of possible samples C({len(population)},{sample_size}) = {n_samples}')

# Compute sample means
sample_means = np.array([np.mean(s) for s in all_samples])

# Display first 10 samples
print(f'\n--- First 10 samples and their means ---')
for i, (s, m) in enumerate(zip(all_samples[:10], sample_means[:10])):
    print(f'  Sample {i+1}: {s}  →  x̄ = {m:.2f}')
print(f'  ...')

# The key result
mean_of_means = sample_means.mean()
print(f'\n✓ Mean of ALL sample means = {mean_of_means:.4f}')
print(f'✓ Population mean μ         = {pop_mean:.4f}')
print(f'✓ They match! E[x̄] = μ  (unbiased estimator)')
```

```python
# --- Plot the sampling distribution ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sampling distribution histogram
axes[0].hist(sample_means, bins=25, color='#4B2E83', edgecolor='white',
             alpha=0.85, density=True)
axes[0].axvline(pop_mean, color='#C62828', lw=2.5, ls='--', label=f'μ = {pop_mean}')
axes[0].axvline(mean_of_means, color='#E8B923', lw=2, ls=':',
                label=f'E[x̄] = {mean_of_means:.2f}')
axes[0].set_title('Sampling Distribution of x̄ (n=3)', fontsize=14, color='#4B2E83')
axes[0].set_xlabel('Sample Mean')
axes[0].set_ylabel('Density')
axes[0].legend(fontsize=11)

# Population vs Sampling distribution comparison
axes[1].hist(population, bins=10, color='#85714D', edgecolor='white',
             alpha=0.6, density=True, label='Population')
axes[1].hist(sample_means, bins=25, color='#4B2E83', edgecolor='white',
             alpha=0.6, density=True, label='Sampling Dist (n=3)')
axes[1].axvline(pop_mean, color='#C62828', lw=2.5, ls='--', label=f'μ = {pop_mean}')
axes[1].set_title('Population vs Sampling Distribution', fontsize=14, color='#4B2E83')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()

# Standard error
se_theory = population.std(ddof=0) / np.sqrt(sample_size)
se_empirical = sample_means.std(ddof=0)
print(f'Standard Error (theory)  = σ/√n = {se_theory:.4f}')
print(f'Standard Error (empirical) = {se_empirical:.4f}')
```

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Aligning Research Questions with Methods</em></p>
</div>

### **OBSERVATIONAL**
- Researcher does **not** manipulate variables
- Observes what naturally occurs
- **Cannot** establish causality
- Example: Surveying smokers vs non-smokers

### **EXPERIMENTAL**
- Researcher **controls** treatment assignment
- **Random assignment** to groups
- Gold standard for causal claims
- Example: A/B test on a website

### **QUASI-EXPERIMENTAL**
- Has a treatment, but **no random assignment**
- Uses natural variation or cutoffs
- Difference-in-differences, regression discontinuity
- Example: Policy change at age 65 (Medicare eligibility)

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Only randomized experiments can establish causation. All other designs require careful assumptions.</p>
</div>

```python
# --- Example: Observational study in pandas ---
np.random.seed(42)
n = 500

# Observational: we simply observe, no randomization
# Confound: people who exercise more also tend to eat healthier
health_consciousness = np.random.normal(0, 1, n)  # hidden confounder

exercise_hours = np.clip(3 + 2 * health_consciousness + np.random.normal(0, 1, n), 0, 15)
healthy_diet_score = np.clip(50 + 10 * health_consciousness + np.random.normal(0, 5, n), 0, 100)
weight = np.clip(180 - 3 * exercise_hours - 0.5 * healthy_diet_score +
                 np.random.normal(0, 10, n), 100, 300)

obs_df = pd.DataFrame({
    'exercise_hours_week': exercise_hours.round(1),
    'healthy_diet_score': healthy_diet_score.round(1),
    'weight_lbs': weight.round(1),
    'study_type': 'observational'
})

print('=== Observational Study ===')
print(obs_df.describe().round(2))
print(f'\nCorrelation (exercise vs weight): {obs_df["exercise_hours_week"].corr(obs_df["weight_lbs"]):.3f}')
print('⚠️  We cannot say exercise CAUSES weight loss — diet is a confounder!')
```

```python
# --- Example: Experimental study (A/B test) ---
np.random.seed(42)
n = 400

# Random assignment to treatment/control
group = np.random.choice(['Control', 'Treatment'], size=n)

# Treatment effect: new website design increases conversion by 5%
base_conversion = 0.10
treatment_effect = 0.05

conversion = np.where(
    group == 'Treatment',
    np.random.binomial(1, base_conversion + treatment_effect, n),
    np.random.binomial(1, base_conversion, n)
)

exp_df = pd.DataFrame({'group': group, 'converted': conversion})

print('=== Experimental Study (A/B Test) ===')
print(exp_df.groupby('group')['converted'].agg(['mean', 'count', 'sum']))
print(f'\nDifference in conversion: '
      f'{exp_df[exp_df.group=="Treatment"]["converted"].mean() - exp_df[exp_df.group=="Control"]["converted"].mean():.3f}')
print('✓ Random assignment → we CAN make causal claims')
```

```python
# --- Example: Quasi-experimental (Regression Discontinuity) ---
np.random.seed(42)
n = 500

# Students with scores >= 70 get a scholarship (the cutoff)
test_score = np.random.uniform(40, 100, n)
scholarship = (test_score >= 70).astype(int)

# GPA is affected by score AND scholarship
gpa = 2.0 + 0.02 * test_score + 0.4 * scholarship + np.random.normal(0, 0.3, n)
gpa = np.clip(gpa, 0, 4.0)

quasi_df = pd.DataFrame({
    'test_score': test_score.round(1),
    'scholarship': scholarship,
    'gpa': gpa.round(2)
})

# Plot regression discontinuity
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['#85714D' if s == 0 else '#4B2E83' for s in scholarship]
ax.scatter(test_score, gpa, c=colors, alpha=0.4, s=20)
ax.axvline(70, color='#C62828', lw=2, ls='--', label='Cutoff (score=70)')

# Fit lines on each side
below = quasi_df[quasi_df.test_score < 70]
above = quasi_df[quasi_df.test_score >= 70]
z1 = np.polyfit(below.test_score, below.gpa, 1)
z2 = np.polyfit(above.test_score, above.gpa, 1)
x1 = np.linspace(40, 70, 50)
x2 = np.linspace(70, 100, 50)
ax.plot(x1, np.polyval(z1, x1), color='#85714D', lw=2.5, label='No Scholarship')
ax.plot(x2, np.polyval(z2, x2), color='#4B2E83', lw=2.5, label='Scholarship')

ax.set_xlabel('Test Score', fontsize=12)
ax.set_ylabel('GPA', fontsize=12)
ax.set_title('Regression Discontinuity Design', fontsize=14, color='#4B2E83')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
print('Quasi-experimental: No random assignment, but the cutoff creates a natural experiment')
```

  <p style="color: #E8D3FF; font-family: 'Encode Sans', 'Arial', sans-serif; font-size: 1.3em; margin: 10px 0; font-weight: 300;">Why 'Correlation Does Not Imply Causation' Matters</p>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">🔗 Correlation vs Causation</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Understanding the Difference</em></p>
</div>

### **CORRELATION**
- X and Y **move together** (positive or negative)
- Measured by Pearson r ∈ [-1, 1]
- Does NOT mean one causes the other

### **CAUSATION**
- X actually **produces** a change in Y
- Requires: temporal precedence, covariation, no alternative explanations
- Only provable with randomized experiments (or clever designs)

### **Three Explanations for Correlation**
1. **X causes Y** (causal)
2. **Y causes X** (reverse causation)
3. **Z causes both** (confounder: X ← Z → Y)

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">With enough variables and time, you can always find a correlation. The question is: does it make sense?</p>
</div>

```python
# --- Correlation vs Causation example ---
np.random.seed(42)
n = 200

# Confounder: summer weather
summer = np.random.normal(0, 1, n)

# Ice cream sales and shark attacks both increase in summer
ice_cream = 50 + 15 * summer + np.random.normal(0, 5, n)
shark_attacks = 3 + 1.5 * summer + np.random.normal(0, 1, n)

r = np.corrcoef(ice_cream, shark_attacks)[0, 1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter: misleading correlation
axes[0].scatter(ice_cream, shark_attacks, alpha=0.5, color='#4B2E83', s=30)
z = np.polyfit(ice_cream, shark_attacks, 1)
axes[0].plot(np.sort(ice_cream), np.polyval(z, np.sort(ice_cream)),
             color='#C62828', lw=2)
axes[0].set_xlabel('Ice Cream Sales ($)', fontsize=12)
axes[0].set_ylabel('Shark Attacks', fontsize=12)
axes[0].set_title(f'Spurious Correlation (r = {r:.3f})', fontsize=14, color='#4B2E83')

# Diagram
axes[1].text(0.5, 0.85, 'Summer (Z)', ha='center', fontsize=16,
             fontweight='bold', color='#C62828')
axes[1].text(0.15, 0.35, 'Ice Cream (X)', ha='center', fontsize=14, color='#4B2E83')
axes[1].text(0.85, 0.35, 'Shark Attacks (Y)', ha='center', fontsize=14, color='#4B2E83')
axes[1].annotate('', xy=(0.2, 0.45), xytext=(0.45, 0.8),
                 arrowprops=dict(arrowstyle='->', color='#C62828', lw=2))
axes[1].annotate('', xy=(0.8, 0.45), xytext=(0.55, 0.8),
                 arrowprops=dict(arrowstyle='->', color='#C62828', lw=2))
axes[1].annotate('', xy=(0.7, 0.35), xytext=(0.3, 0.35),
                 arrowprops=dict(arrowstyle='<->', color='gray', lw=1.5, ls='dashed'))
axes[1].text(0.5, 0.25, 'Spurious!', ha='center', fontsize=12,
             color='gray', style='italic')
axes[1].set_xlim(0, 1)
axes[1].set_ylim(0, 1)
axes[1].axis('off')
axes[1].set_title('Confounding Structure: X ← Z → Y', fontsize=14, color='#4B2E83')

plt.tight_layout()
plt.show()
```

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>A Clever 'Natural Experiment'</em></p>
</div>

### **What is an Instrumental Variable?**
- An IV (Z) affects Y **only through** X
- Z → X → Y (Z does NOT directly affect Y)
- Requirements: (1) Relevance: Z is correlated with X, (2) Exogeneity: Z is uncorrelated with the error

### **Classic IV Examples**
- Draft lottery → military service → earnings (Angrist, 1990)
- Rainfall → economic growth → civil conflict
- Quarter of birth → years of schooling → wages
- Distance to college → education → earnings

### **Modelling Confounders**
- Include the confounder as a control variable in regression
- If the confounder is unobserved, use IV or other techniques
- Simple approach: add suspected confounders to the model

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">An instrumental variable is a clever 'natural experiment' — it nudges X without directly touching Y.</p>
</div>

```python
# --- IV Example: Distance to college → Education → Earnings ---
from sklearn.linear_model import LinearRegression
np.random.seed(42)
n = 1000

# Unobserved ability (confounder)
ability = np.random.normal(0, 1, n)

# Instrument: distance to college (miles)
distance = np.random.uniform(0, 100, n)

# Education (years) depends on distance AND ability
education = 16 - 0.03 * distance + 1.5 * ability + np.random.normal(0, 1, n)

# Earnings depend on education AND ability (ability is the confounder)
earnings = 20000 + 3000 * education + 5000 * ability + np.random.normal(0, 5000, n)

# Naive OLS (biased because ability is omitted)
naive = LinearRegression().fit(education.reshape(-1,1), earnings)
print(f'Naive OLS coefficient (education → earnings): ${naive.coef_[0]:,.0f}')
print('  ⚠️ Biased upward because ability is omitted!\n')

# 2-Stage Least Squares (IV)
# Stage 1: Regress education on distance (the instrument)
stage1 = LinearRegression().fit(distance.reshape(-1,1), education)
education_hat = stage1.predict(distance.reshape(-1,1))
print(f'Stage 1: distance → education (coef = {stage1.coef_[0]:.4f})')

# Stage 2: Regress earnings on predicted education
stage2 = LinearRegression().fit(education_hat.reshape(-1,1), earnings)
print(f'Stage 2 (IV estimate): education → earnings = ${stage2.coef_[0]:,.0f}')
print(f'True effect: $3,000')
print(f'\n✓ The IV estimate is closer to the true causal effect!')
```

```python
# --- Modelling confounders directly ---
np.random.seed(42)
n = 500

# Confounder: shoe size and math score both depend on age
age = np.random.uniform(5, 18, n)
shoe_size = 3 + 0.5 * age + np.random.normal(0, 1, n)
math_score = 20 + 5 * age + np.random.normal(0, 10, n)

# Without controlling for age
r_naive = np.corrcoef(shoe_size, math_score)[0,1]

# Controlling for age (partial correlation via residuals)
shoe_resid = shoe_size - LinearRegression().fit(age.reshape(-1,1), shoe_size).predict(age.reshape(-1,1))
math_resid = math_score - LinearRegression().fit(age.reshape(-1,1), math_score).predict(age.reshape(-1,1))
r_partial = np.corrcoef(shoe_resid, math_resid)[0,1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(shoe_size, math_score, alpha=0.4, c='#4B2E83', s=20)
axes[0].set_title(f'Naive: Shoe Size vs Math Score (r={r_naive:.3f})',
                  fontsize=13, color='#4B2E83')
axes[0].set_xlabel('Shoe Size')
axes[0].set_ylabel('Math Score')

axes[1].scatter(shoe_resid, math_resid, alpha=0.4, c='#E8B923', s=20)
axes[1].set_title(f'After Controlling for Age (r={r_partial:.3f})',
                  fontsize=13, color='#4B2E83')
axes[1].set_xlabel('Shoe Size (residual)')
axes[1].set_ylabel('Math Score (residual)')

plt.tight_layout()
plt.show()
print(f'Naive correlation: {r_naive:.3f}')
print(f'Partial correlation (controlling for age): {r_partial:.3f}')
print('The spurious correlation vanishes when we control for the confounder!')
```

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">🔀 Mediators & Moderators</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Understanding How and When Effects Occur</em></p>
</div>

### **MEDIATOR (the HOW)**
- Explains the **mechanism** through which X affects Y
- Path: X → M → Y
- Example: Exercise → Fitness → Weight Loss
- Example: Education → Skills → Income

### **MODERATOR (the WHEN)**
- Changes the **strength/direction** of the X → Y relationship
- X → Y depends on the level of W
- Example: Coffee → Productivity, moderated by sleep quality
- Example: Tutoring → Grades, moderated by motivation

### **Path Models in semopy**
- semopy uses lavaan-like syntax to define structural models
- `y ~ x` means "y is regressed on x"
- `y =~ x1 + x2` means "y is a latent variable measured by x1 and x2"
- Models can include mediators, moderators, and latent variables

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Mediator = HOW the effect happens. Moderator = WHEN or FOR WHOM the effect changes.</p>
</div>

```python
# --- Example 1: Simple Mediation (X → M → Y) ---
import semopy
np.random.seed(42)
n = 500

# Generate data: Education → Skills → Income
education = np.random.normal(14, 3, n)
skills = 2 + 0.8 * education + np.random.normal(0, 2, n)
income = 15000 + 500 * skills + 200 * education + np.random.normal(0, 5000, n)

data = pd.DataFrame({'education': education, 'skills': skills, 'income': income})

# Define the mediation model
model_spec = """
skills ~ education
income ~ skills + education
"""

model = semopy.Model(model_spec)
result = model.fit(data)
estimates = model.inspect()
print('=== Example 1: Simple Mediation ===')
print('Path: Education → Skills → Income')
print(estimates[['lval', 'op', 'rval', 'Estimate', 'p-value']].to_string(index=False))

# Compute indirect effect
a = estimates[(estimates.lval=='skills') & (estimates.rval=='education')]['Estimate'].values[0]
b = estimates[(estimates.lval=='income') & (estimates.rval=='skills')]['Estimate'].values[0]
c_prime = estimates[(estimates.lval=='income') & (estimates.rval=='education')]['Estimate'].values[0]
print(f'\nDirect effect (education → income): {c_prime:.2f}')
print(f'Indirect effect (education → skills → income): {a:.2f} × {b:.2f} = {a*b:.2f}')
print(f'Total effect: {c_prime + a*b:.2f}')
```

```python
# --- Example 2: Removing spurious correlation by modelling a confounder ---
np.random.seed(42)
n = 500

# Confounder: summer temperature
temperature = np.random.normal(75, 10, n)
ice_cream = 5 + 0.3 * temperature + np.random.normal(0, 3, n)
drownings = 0.5 + 0.05 * temperature + np.random.normal(0, 1, n)

data2 = pd.DataFrame({'temperature': temperature, 'ice_cream': ice_cream,
                       'drownings': drownings})

# Naive correlation
print(f'Naive correlation (ice_cream, drownings): {data2["ice_cream"].corr(data2["drownings"]):.3f}')

# SEM model with confounder
model_spec2 = """
ice_cream ~ temperature
drownings ~ temperature + ice_cream
"""

model2 = semopy.Model(model_spec2)
model2.fit(data2)
est2 = model2.inspect()
print('\n=== SEM: Controlling for Temperature ===')
print(est2[['lval', 'op', 'rval', 'Estimate', 'p-value']].to_string(index=False))
print('\n✓ Ice cream → drownings becomes non-significant when temperature is controlled!')
```

```python
# --- Example 3: Moderation (interaction effect) ---
np.random.seed(42)
n = 500

# Coffee → Productivity, moderated by Sleep Quality
coffee = np.random.uniform(0, 5, n)  # cups per day
sleep_quality = np.random.uniform(1, 10, n)  # 1-10 scale
interaction = coffee * sleep_quality

# Productivity depends on coffee, sleep, and their interaction
productivity = (50 + 3 * coffee + 2 * sleep_quality + 1.5 * interaction
                + np.random.normal(0, 10, n))

data3 = pd.DataFrame({
    'coffee': coffee, 'sleep_quality': sleep_quality,
    'interaction': interaction, 'productivity': productivity
})

model_spec3 = """
productivity ~ coffee + sleep_quality + interaction
"""

model3 = semopy.Model(model_spec3)
model3.fit(data3)
est3 = model3.inspect()
print('=== Example 3: Moderation ===')
print('Coffee → Productivity, moderated by Sleep Quality')
print(est3[['lval', 'op', 'rval', 'Estimate', 'p-value']].to_string(index=False))

# Visualize moderation
fig, ax = plt.subplots(figsize=(10, 6))
for sq, color, label in [(2, '#C62828', 'Low Sleep'), (5, '#E8B923', 'Med Sleep'), (8, '#4B2E83', 'High Sleep')]:
    mask = (sleep_quality > sq - 1.5) & (sleep_quality < sq + 1.5)
    ax.scatter(coffee[mask], productivity[mask], alpha=0.3, c=color, s=20)
    z = np.polyfit(coffee[mask], productivity[mask], 1)
    x_line = np.linspace(0, 5, 50)
    ax.plot(x_line, np.polyval(z, x_line), color=color, lw=2.5, label=f'{label} (slope={z[0]:.1f})')

ax.set_xlabel('Coffee (cups/day)', fontsize=12)
ax.set_ylabel('Productivity', fontsize=12)
ax.set_title('Moderation: Sleep Quality Changes the Coffee→Productivity Effect',
             fontsize=14, color='#4B2E83')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# --- Example 4: Full path model with latent variables ---
np.random.seed(42)
n = 500

# Simulate: Ads → Brand Awareness → Purchase Intent → Sales
ads = np.random.normal(100, 20, n)
brand_awareness = 20 + 0.3 * ads + np.random.normal(0, 10, n)
purchase_intent = 10 + 0.5 * brand_awareness + 0.1 * ads + np.random.normal(0, 8, n)
sales = 500 + 5 * purchase_intent + 1 * brand_awareness + np.random.normal(0, 50, n)

data4 = pd.DataFrame({
    'ads': ads, 'brand_awareness': brand_awareness,
    'purchase_intent': purchase_intent, 'sales': sales
})

model_spec4 = """
brand_awareness ~ ads
purchase_intent ~ brand_awareness + ads
sales ~ purchase_intent + brand_awareness
"""

model4 = semopy.Model(model_spec4)
model4.fit(data4)
est4 = model4.inspect()
print('=== Example 4: Full Path Model ===')
print('Ads → Brand Awareness → Purchase Intent → Sales')
print(est4[['lval', 'op', 'rval', 'Estimate', 'p-value']].to_string(index=False))

# Model fit statistics
fit_stats = semopy.calc_stats(model4)
print(f'\nModel Fit Statistics:')
print(fit_stats.T)
```

### **What is Canonical Correlation?**
- Finds linear combinations of two variable sets that are maximally correlated
- Set 1: {reading, math, science} → Set 2: {study_hours, sleep, motivation}
- Extends correlation to **multivariate** settings
- Number of canonical correlations = min(p, q)

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">Canonical Correlation reveals hidden relationships between two groups of variables simultaneously.</p>
</div>

```python
# --- Canonical Correlation Analysis ---
from sklearn.cross_decomposition import CCA
np.random.seed(42)
n = 200

# Academic scores (Set X)
study_effort = np.random.normal(0, 1, n)
reading = 60 + 10 * study_effort + np.random.normal(0, 5, n)
math = 65 + 12 * study_effort + np.random.normal(0, 6, n)
science = 58 + 11 * study_effort + np.random.normal(0, 7, n)

# Lifestyle (Set Y)
study_hours = 3 + 2 * study_effort + np.random.normal(0, 1, n)
sleep = 7 + 0.5 * study_effort + np.random.normal(0, 0.5, n)
motivation = 5 + 1.5 * study_effort + np.random.normal(0, 1, n)

X = np.column_stack([reading, math, science])
Y = np.column_stack([study_hours, sleep, motivation])

# Fit CCA
cca = CCA(n_components=2)
X_c, Y_c = cca.fit_transform(X, Y)

# Canonical correlations
for i in range(2):
    r = np.corrcoef(X_c[:, i], Y_c[:, i])[0, 1]
    print(f'Canonical Correlation {i+1}: r = {r:.4f}')

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X_c[:, 0], Y_c[:, 0], alpha=0.5, c='#4B2E83', s=25)
axes[0].set_xlabel('Canonical Variate X₁', fontsize=12)
axes[0].set_ylabel('Canonical Variate Y₁', fontsize=12)
axes[0].set_title(f'1st Canonical Correlation (r={np.corrcoef(X_c[:,0], Y_c[:,0])[0,1]:.3f})',
                  fontsize=13, color='#4B2E83')

axes[1].scatter(X_c[:, 1], Y_c[:, 1], alpha=0.5, c='#E8B923', s=25)
axes[1].set_xlabel('Canonical Variate X₂', fontsize=12)
axes[1].set_ylabel('Canonical Variate Y₂', fontsize=12)
axes[1].set_title(f'2nd Canonical Correlation (r={np.corrcoef(X_c[:,1], Y_c[:,1])[0,1]:.3f})',
                  fontsize=13, color='#4B2E83')

plt.suptitle('Canonical Correlation Analysis', fontsize=15, color='#4B2E83', y=1.02)
plt.tight_layout()
plt.show()
```

### **What is MANOVA?**
- Extension of ANOVA to **multiple dependent variables** simultaneously
- Tests whether groups differ on a combination of DVs
- Example: Do 3 teaching methods differ on {test_score, satisfaction, retention}?
- Uses Wilks' Lambda, Pillai's Trace, or Hotelling's T²

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">MANOVA tests whether groups differ on several dependent variables simultaneously, accounting for correlations among DVs.</p>
</div>

```python
# --- MANOVA Example: Teaching Methods ---
from scipy.stats import f_oneway
np.random.seed(42)

# Three teaching methods, three outcomes
n_per = 50
methods = np.repeat(['Lecture', 'Hands-on', 'Online'], n_per)

# Generate correlated outcomes per group
data_groups = []
means = {'Lecture': [70, 3.5, 60], 'Hands-on': [78, 4.2, 75], 'Online': [72, 3.8, 65]}
for method in ['Lecture', 'Hands-on', 'Online']:
    mu = means[method]
    scores = np.random.multivariate_normal(
        mu, [[100, 2, 30], [2, 0.5, 1], [30, 1, 80]], n_per)
    data_groups.append(scores)

all_data = np.vstack(data_groups)
manova_df = pd.DataFrame(all_data, columns=['test_score', 'satisfaction', 'retention'])
manova_df['method'] = methods

print('=== Group Means ===')
print(manova_df.groupby('method')[['test_score', 'satisfaction', 'retention']].mean().round(2))

# MANOVA using statsmodels
try:
    from statsmodels.multivariate.manova import MANOVA
    manova = MANOVA.from_formula('test_score + satisfaction + retention ~ method', data=manova_df)
    print('\n=== MANOVA Results ===')
    print(manova.mv_test())
except ImportError:
    print('\nstatsmodels not available. Individual ANOVAs:')
    for col in ['test_score', 'satisfaction', 'retention']:
        groups = [manova_df[manova_df.method==m][col] for m in ['Lecture', 'Hands-on', 'Online']]
        f, p = f_oneway(*groups)
        print(f'  {col}: F={f:.2f}, p={p:.4f}')

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(['test_score', 'satisfaction', 'retention']):
    manova_df.boxplot(column=col, by='method', ax=axes[i])
    axes[i].set_title(col.replace('_', ' ').title(), fontsize=13, color='#4B2E83')
    axes[i].set_xlabel('')
plt.suptitle('MANOVA: Teaching Methods', fontsize=15, color='#4B2E83', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# --- Multiple Regression Example ---
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
np.random.seed(42)

# Simulated house price data
n = 300
size_sqft = np.random.uniform(800, 3500, n)
bedrooms = np.random.choice([1,2,3,4,5], n, p=[0.05,0.15,0.4,0.3,0.1])
age = np.random.uniform(0, 50, n)

price = (50000 + 150 * size_sqft + 20000 * bedrooms - 1000 * age
         + np.random.normal(0, 30000, n))

reg_df = pd.DataFrame({
    'size_sqft': size_sqft, 'bedrooms': bedrooms,
    'age': age, 'price': price
})

# Fit regression
X = reg_df[['size_sqft', 'bedrooms', 'age']]
y = reg_df['price']
model = LinearRegression().fit(X, y)

print('=== Multiple Regression: House Prices ===')
print(f'Intercept: ${model.intercept_:,.0f}')
for name, coef in zip(X.columns, model.coef_):
    print(f'  {name}: ${coef:,.0f}')
print(f'R² = {model.score(X, y):.4f}')

# Plot actual vs predicted
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

y_pred = model.predict(X)
axes[0].scatter(y, y_pred, alpha=0.4, c='#4B2E83', s=20)
lims = [min(y.min(), y_pred.min()), max(y.max(), y_pred.max())]
axes[0].plot(lims, lims, 'r--', lw=2)
axes[0].set_xlabel('Actual Price ($)', fontsize=12)
axes[0].set_ylabel('Predicted Price ($)', fontsize=12)
axes[0].set_title('Actual vs Predicted', fontsize=14, color='#4B2E83')

# Residual plot
residuals = y - y_pred
axes[1].scatter(y_pred, residuals, alpha=0.4, c='#E8B923', s=20)
axes[1].axhline(0, color='red', lw=2, ls='--')
axes[1].set_xlabel('Predicted Price ($)', fontsize=12)
axes[1].set_ylabel('Residual', fontsize=12)
axes[1].set_title('Residual Plot', fontsize=14, color='#4B2E83')

plt.tight_layout()
plt.show()
```

```python
# --- Conjoint Analysis Example ---
np.random.seed(42)

# Phone features and their levels
prices = [499, 699, 999]
screens = [5.5, 6.1, 6.7]  # inches
batteries = [3000, 4000, 5000]  # mAh

# Full factorial design
from itertools import product as iproduct
profiles = list(iproduct(prices, screens, batteries))
n_profiles = len(profiles)
n_respondents = 100

# Simulate preferences (utility = -0.005*price + 2*screen + 0.002*battery + noise)
results = []
for resp in range(n_respondents):
    for price, screen, battery in profiles:
        utility = (-0.005 * price + 2 * screen + 0.002 * battery
                   + np.random.normal(0, 1))
        results.append([resp, price, screen, battery, utility])

conj_df = pd.DataFrame(results, columns=['respondent', 'price', 'screen', 'battery', 'preference'])

# Fit regression to estimate part-worths
X_conj = conj_df[['price', 'screen', 'battery']]
y_conj = conj_df['preference']
conj_model = LinearRegression().fit(X_conj, y_conj)

print('=== Conjoint Analysis: Phone Preferences ===')
print(f'Part-worth utilities (regression coefficients):')
for name, coef in zip(['price', 'screen', 'battery'], conj_model.coef_):
    print(f'  {name}: {coef:.4f}')
print(f'R² = {conj_model.score(X_conj, y_conj):.4f}')

# Relative importance
ranges = {
    'price': abs(conj_model.coef_[0]) * (max(prices) - min(prices)),
    'screen': abs(conj_model.coef_[1]) * (max(screens) - min(screens)),
    'battery': abs(conj_model.coef_[2]) * (max(batteries) - min(batteries))
}
total_range = sum(ranges.values())
importance = {k: v/total_range*100 for k, v in ranges.items()}

fig, ax = plt.subplots(figsize=(8, 5))
colors = ['#4B2E83', '#E8B923', '#32006E']
ax.bar(importance.keys(), importance.values(), color=colors, edgecolor='white')
ax.set_ylabel('Relative Importance (%)', fontsize=12)
ax.set_title('Feature Importance in Phone Purchase', fontsize=14, color='#4B2E83')
for i, (k, v) in enumerate(importance.items()):
    ax.text(i, v + 1, f'{v:.1f}%', ha='center', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# --- Discriminant Analysis Example ---
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
np.random.seed(42)

# Loan default classification
n = 400
income = np.concatenate([np.random.normal(60, 15, 300), np.random.normal(40, 10, 100)])
credit_score = np.concatenate([np.random.normal(700, 50, 300), np.random.normal(580, 40, 100)])
default = np.concatenate([np.zeros(300), np.ones(100)])

X_disc = np.column_stack([income, credit_score])
X_train, X_test, y_train, y_test = train_test_split(X_disc, default, test_size=0.3, random_state=42)

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)

print('=== Linear Discriminant Analysis: Loan Default ===')
print(classification_report(y_test, y_pred, target_names=['No Default', 'Default']))

# Decision boundary plot
fig, ax = plt.subplots(figsize=(10, 7))
xx, yy = np.meshgrid(np.linspace(income.min()-5, income.max()+5, 200),
                      np.linspace(credit_score.min()-20, credit_score.max()+20, 200))
Z = lda.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
ax.contourf(xx, yy, Z, alpha=0.15, cmap='RdPu')
ax.contour(xx, yy, Z, colors='#4B2E83', linewidths=1.5)

scatter = ax.scatter(income, credit_score, c=default, cmap='RdPu',
                     alpha=0.6, s=30, edgecolors='white', linewidths=0.5)
ax.set_xlabel('Income ($K)', fontsize=12)
ax.set_ylabel('Credit Score', fontsize=12)
ax.set_title('LDA Decision Boundary: Loan Default', fontsize=14, color='#4B2E83')
ax.legend(*scatter.legend_elements(), labels=['No Default', 'Default'], fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# --- Linear Probability Model ---
np.random.seed(42)
n = 500

age = np.random.uniform(18, 80, n)
income = np.random.lognormal(10.5, 0.5, n)

# True probability of voting
prob = 1 / (1 + np.exp(-(  -3 + 0.04 * age + 0.00001 * income  )))
voted = np.random.binomial(1, prob)

lpm_df = pd.DataFrame({'age': age, 'income': income, 'voted': voted})

# LPM (just OLS on binary outcome)
from sklearn.linear_model import LinearRegression
X_lpm = lpm_df[['age', 'income']]
lpm = LinearRegression().fit(X_lpm, voted)

print('=== Linear Probability Model: Voting ===')
print(f'Intercept: {lpm.intercept_:.4f}')
print(f'Age coefficient: {lpm.coef_[0]:.5f}  (each year → +{lpm.coef_[0]*100:.2f}pp)')
print(f'Income coefficient: {lpm.coef_[1]:.8f}')
print(f'R² = {lpm.score(X_lpm, voted):.4f}')

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# LPM predictions vs age
age_grid = np.linspace(18, 80, 100)
income_mean = income.mean()
pred_grid = lpm.predict(np.column_stack([age_grid, np.full(100, income_mean)]))

axes[0].scatter(age, voted, alpha=0.15, c='#4B2E83', s=15)
axes[0].plot(age_grid, pred_grid, color='#C62828', lw=2.5, label='LPM fit')
axes[0].set_xlabel('Age', fontsize=12)
axes[0].set_ylabel('P(Voted)', fontsize=12)
axes[0].set_title('LPM: Age vs Voting Probability', fontsize=14, color='#4B2E83')
axes[0].set_ylim(-0.1, 1.1)
axes[0].legend(fontsize=11)

# Predicted probabilities distribution
all_preds = lpm.predict(X_lpm)
axes[1].hist(all_preds, bins=30, color='#4B2E83', edgecolor='white', alpha=0.8)
axes[1].axvline(0, color='red', ls='--', lw=1.5, label='P=0')
axes[1].axvline(1, color='red', ls='--', lw=1.5, label='P=1')
axes[1].set_xlabel('Predicted Probability', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].set_title('LPM Predicted Probabilities', fontsize=14, color='#4B2E83')
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()
print(f'\n⚠️  LPM can predict probabilities < 0 or > 1!')
print(f'   Predictions < 0: {(all_preds < 0).sum()}, Predictions > 1: {(all_preds > 1).sum()}')
```

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>Validating Measurement Models</em></p>
</div>

### **What is CFA?**
- Tests whether observed variables load onto **pre-specified latent factors**
- Unlike EFA (exploratory), CFA **confirms** a theoretical structure
- Example: Do 10 survey items load onto 3 constructs as hypothesized?
- Uses fit indices: CFI, TLI, RMSEA, SRMR

<div style="background-color: #E8D3FF; border-left: 4px solid #4B2E83; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #4B2E83; font-weight: 600;">💡 Key Principle:</p>
  <p style="margin: 10px 0 0 0; color: #333;">CFA validates that your measurement instrument measures what it claims to measure.</p>
</div>

```python
# --- Confirmatory Factor Analysis with semopy ---
import semopy
np.random.seed(42)
n = 300

# Two latent factors: Math Ability and Verbal Ability
math_ability = np.random.normal(0, 1, n)
verbal_ability = np.random.normal(0, 1, n)

# Observed indicators
math1 = 50 + 10 * math_ability + np.random.normal(0, 3, n)
math2 = 55 + 8 * math_ability + np.random.normal(0, 4, n)
math3 = 48 + 12 * math_ability + np.random.normal(0, 3.5, n)

verbal1 = 60 + 9 * verbal_ability + np.random.normal(0, 3, n)
verbal2 = 58 + 11 * verbal_ability + np.random.normal(0, 4, n)
verbal3 = 62 + 7 * verbal_ability + np.random.normal(0, 3, n)

cfa_data = pd.DataFrame({
    'math1': math1, 'math2': math2, 'math3': math3,
    'verbal1': verbal1, 'verbal2': verbal2, 'verbal3': verbal3
})

# CFA model specification
cfa_spec = """
MathAbility =~ math1 + math2 + math3
VerbalAbility =~ verbal1 + verbal2 + verbal3
"""

cfa_model = semopy.Model(cfa_spec)
cfa_model.fit(cfa_data)
cfa_est = cfa_model.inspect()

print('=== Confirmatory Factor Analysis ===')
print('All Estimates:')
print(cfa_est.to_string(index=False))

# Extract factor loadings (operator '=~')
loadings = cfa_est[cfa_est['op'] == '=~'][['lval', 'rval', 'Estimate']].copy()
if loadings.empty:
    # Some semopy versions use '~' for all relationships
    loadings = cfa_est[cfa_est['rval'].isin(
        ['math1','math2','math3','verbal1','verbal2','verbal3']
    )][['lval', 'rval', 'Estimate']].copy()

print('\nFactor Loadings:')
print(loadings.to_string(index=False))

# Fit statistics
try:
    fit = semopy.calc_stats(cfa_model)
    print(f'\nModel Fit:')
    print(fit.T)
except Exception as e:
    print(f'\nFit stats not available: {e}')

# Visualization: factor loadings bar chart
fig, ax = plt.subplots(figsize=(10, 5))
if not loadings.empty:
    n_math = min(3, len(loadings))
    colors = ['#4B2E83'] * n_math + ['#E8B923'] * (len(loadings) - n_math)
    loadings_vals = loadings['Estimate'].values.astype(float)
    labels = [f"{r['lval']} → {r['rval']}" for _, r in loadings.iterrows()]
    ax.barh(labels, loadings_vals, color=colors[:len(labels)], edgecolor='white')
    ax.set_xlabel('Loading', fontsize=12)
    ax.set_title('CFA Factor Loadings', fontsize=14, color='#4B2E83')
    ax.axvline(0, color='gray', lw=0.5)
else:
    ax.text(0.5, 0.5, 'Loadings not extracted — check semopy output above',
            ha='center', va='center', transform=ax.transAxes, fontsize=12)
    ax.set_title('CFA Factor Loadings', fontsize=14, color='#4B2E83')
plt.tight_layout()
plt.show()
```

```python
# --- Cluster Analysis: Customer Segmentation ---
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
np.random.seed(42)

# Customer data with 3 natural segments
n_cust = 300
seg1 = np.random.multivariate_normal([30, 200, 2], [[25,0,0],[0,2500,0],[0,0,1]], 100)
seg2 = np.random.multivariate_normal([50, 800, 8], [[30,0,0],[0,4000,0],[0,0,4]], 100)
seg3 = np.random.multivariate_normal([65, 1500, 15], [[20,0,0],[0,5000,0],[0,0,6]], 100)

data_clust = np.vstack([seg1, seg2, seg3])
clust_df = pd.DataFrame(data_clust, columns=['age', 'spending', 'frequency'])

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(clust_df)

# K-Means with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clust_df['cluster'] = kmeans.fit_predict(X_scaled)

print('=== Cluster Analysis: Customer Segments ===')
print(clust_df.groupby('cluster')[['age', 'spending', 'frequency']].mean().round(1))

# Elbow plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

inertias = []
K_range = range(1, 8)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_scaled)
    inertias.append(km.inertia_)

axes[0].plot(K_range, inertias, 'o-', color='#4B2E83', lw=2, ms=8)
axes[0].set_xlabel('k (number of clusters)', fontsize=12)
axes[0].set_ylabel('Inertia', fontsize=12)
axes[0].set_title('Elbow Method', fontsize=14, color='#4B2E83')

# Scatter plot
colors_map = {0: '#4B2E83', 1: '#E8B923', 2: '#C62828'}
for c in [0, 1, 2]:
    mask = clust_df.cluster == c
    axes[1].scatter(clust_df[mask].spending, clust_df[mask].frequency,
                    c=colors_map[c], alpha=0.6, s=30, label=f'Cluster {c}')
axes[1].set_xlabel('Spending ($)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title('Customer Segments', fontsize=14, color='#4B2E83')
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# --- Multidimensional Scaling: Car Brand Perceptions ---
from sklearn.manifold import MDS
np.random.seed(42)

brands = ['Toyota', 'Honda', 'BMW', 'Mercedes', 'Ford',
          'Tesla', 'Hyundai', 'Audi', 'Kia', 'Lexus']

# Simulated similarity matrix (lower = more similar)
# Luxury brands cluster, economy brands cluster
n_brands = len(brands)
similarity = np.random.uniform(3, 8, (n_brands, n_brands))
similarity = (similarity + similarity.T) / 2  # symmetrize
np.fill_diagonal(similarity, 0)

# Make luxury brands similar to each other
luxury = [2, 3, 7, 9]  # BMW, Mercedes, Audi, Lexus
economy = [0, 1, 4, 6, 8]  # Toyota, Honda, Ford, Hyundai, Kia
for i in luxury:
    for j in luxury:
        if i != j:
            val = np.random.uniform(1, 3)
            similarity[i, j] = val
            similarity[j, i] = val
for i in economy:
    for j in economy:
        if i != j:
            val = np.random.uniform(1, 3)
            similarity[i, j] = val
            similarity[j, i] = val
# Tesla is in between
for idx in luxury:
    val = np.random.uniform(2, 5)
    similarity[5, idx] = val
    similarity[idx, 5] = val

# Final symmetry check
np.fill_diagonal(similarity, 0)

# MDS
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42, normalized_stress='auto')
coords = mds.fit_transform(similarity)

# Plot perceptual map
from matplotlib.patches import Patch
fig, ax = plt.subplots(figsize=(10, 8))
colors = ['#E8B923' if i in luxury else '#4B2E83' for i in range(n_brands)]
colors[5] = '#C62828'  # Tesla special

ax.scatter(coords[:, 0], coords[:, 1], c=colors, s=150, zorder=5, edgecolors='white', linewidths=1.5)
for i, brand in enumerate(brands):
    ax.annotate(brand, (coords[i, 0], coords[i, 1]),
                textcoords='offset points', xytext=(10, 5),
                fontsize=12, fontweight='bold', color=colors[i])

ax.set_xlabel('Dimension 1', fontsize=12)
ax.set_ylabel('Dimension 2', fontsize=12)
ax.set_title('MDS: Car Brand Perceptual Map', fontsize=14, color='#4B2E83')
ax.axhline(0, color='gray', lw=0.5, ls=':')
ax.axvline(0, color='gray', lw=0.5, ls=':')

# Add legend
legend_elements = [
    Patch(facecolor='#E8B923', label='Luxury'),
    Patch(facecolor='#4B2E83', label='Economy'),
    Patch(facecolor='#C62828', label='Tesla (hybrid)')
]
ax.legend(handles=legend_elements, fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# --- Correspondence Analysis: Job Roles × Skills ---
from matplotlib.patches import Patch
np.random.seed(42)

# Contingency table: job roles x skills
data_ca = pd.DataFrame({
    'Python': [80, 30, 50, 10, 60],
    'SQL': [60, 70, 40, 20, 30],
    'Excel': [20, 60, 80, 90, 15],
    'ML': [70, 10, 30, 5, 80],
    'Communication': [30, 50, 70, 85, 40]
}, index=['Data Scientist', 'Analyst', 'Manager', 'Admin', 'ML Engineer'])

print('=== Contingency Table: Job Roles × Skills ===')
print(data_ca)

# Simple correspondence analysis using SVD
# Normalize the table
P = data_ca.values / data_ca.values.sum()
row_sums = P.sum(axis=1, keepdims=True)
col_sums = P.sum(axis=0, keepdims=True)

# Standardized residuals
S = (P - row_sums @ col_sums) / np.sqrt(row_sums @ col_sums)

# SVD
U, sigma, Vt = np.linalg.svd(S, full_matrices=False)

# Row and column coordinates in 2D
row_coords = (U[:, :2] * sigma[:2]) / np.sqrt(row_sums.flatten()[:, None])
col_coords = (Vt[:2, :].T * sigma[:2]) / np.sqrt(col_sums.flatten()[:, None])

# Inertia explained
inertia = sigma**2
pct_explained = inertia / inertia.sum() * 100

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot rows (job roles)
for i, role in enumerate(data_ca.index):
    ax.scatter(row_coords[i, 0], row_coords[i, 1], c='#4B2E83', s=120, zorder=5)
    ax.annotate(role, (row_coords[i, 0], row_coords[i, 1]),
                textcoords='offset points', xytext=(8, 8),
                fontsize=11, fontweight='bold', color='#4B2E83')

# Plot columns (skills)
for j, skill in enumerate(data_ca.columns):
    ax.scatter(col_coords[j, 0], col_coords[j, 1], c='#E8B923', s=120,
              marker='s', zorder=5)
    ax.annotate(skill, (col_coords[j, 0], col_coords[j, 1]),
                textcoords='offset points', xytext=(8, 8),
                fontsize=11, fontweight='bold', color='#E8B923')

ax.axhline(0, color='gray', lw=0.5, ls=':')
ax.axvline(0, color='gray', lw=0.5, ls=':')
ax.set_xlabel(f'Dimension 1 ({pct_explained[0]:.1f}%)', fontsize=12)
ax.set_ylabel(f'Dimension 2 ({pct_explained[1]:.1f}%)', fontsize=12)
ax.set_title('Correspondence Analysis: Job Roles × Skills', fontsize=14, color='#4B2E83')

legend_elements = [
    Patch(facecolor='#4B2E83', label='Job Roles'),
    Patch(facecolor='#E8B923', label='Skills')
]
ax.legend(handles=legend_elements, fontsize=11)
plt.tight_layout()
plt.show()

print(f'\nInertia explained: Dim1={pct_explained[0]:.1f}%, Dim2={pct_explained[1]:.1f}%')
```
