<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Choosing the Right Chart Type</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Every chart type answers a different question. Matching the chart to the question is the single most important decision you make as a data visualiser. Choosing wrong — e.g., a pie chart for trend data — can actively mislead readers.</p>
</div>

| Chart Type | Best For | Avoid When |
|:-----------|:---------|:-----------|
| **Bar chart** | Comparing discrete categories | You have more than ~15 categories |
| **Horizontal bar** | Long category labels; ranking | Showing change over time |
| **Line chart** | Trends over a continuous axis (time, temperature) | Data points are not connected logically |
| **Scatter plot** | Relationship / correlation between two numeric variables | One variable is categorical |
| **Histogram** | Distribution of a single numeric variable | You want to compare across groups (use boxplot instead) |
| **Box plot** | Distribution comparison across groups; spotting outliers | Your audience is unfamiliar with the chart |
| **Pie / donut** | Part-of-whole with **≤ 5** clearly different slices | You have many categories or need precision |
| **Heatmap** | Matrix-shaped relationships (correlation, confusion matrix) | The matrix is too sparse |

> **Rule of thumb:** if you cannot state in one sentence what question your chart answers, redesign it.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Readable Titles and Subtitles</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A title should answer <strong>"what is the main finding?"</strong> — not just label the data. Compare these two titles for the same chart:</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li style="color: #C62828;"><strong>Weak:</strong> &ldquo;Sales Data 2023&rdquo; — describes the data, not the insight.</li>
    <li style="color: #388E3C;"><strong>Strong:</strong> &ldquo;Q4 Online Sales Surged 38% Above the Annual Average&rdquo; — states the finding.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7; margin-top: 12px;">Use a <strong>subtitle</strong> (smaller, lighter text below the title) to provide context: data source, time period, unit of measurement, or a brief caveat. In matplotlib use <code>fig.suptitle()</code> for the main title and <code>ax.set_title()</code> for a panel subtitle, or compose both with <code>ax.set_title()</code> and an annotation.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Axis Labels and Units</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Every axis that carries quantitative information <strong>must be labelled with the variable name and its unit</strong>. Omitting units is one of the most common and costly mistakes in data communication. A reader who sees &ldquo;47&rdquo; on a y-axis cannot know whether that means dollars, thousands of dollars, kilograms, or minutes.</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li>Use plain language: <code>Revenue (USD millions)</code> not <code>rev_usd_m</code>.</li>
    <li>Keep axis tick labels horizontal wherever possible — rotate only when unavoidable, and never past 45°.</li>
    <li>Do not truncate the y-axis to exaggerate differences unless you explicitly note the broken axis.</li>
    <li>Match the scale to the data: avoid vast empty white space at the top of a bar chart.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Legends and Direct Labels</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A <strong>legend</strong> forces the reader's eye to shuttle back and forth between the chart and the key — each trip costs attention. Wherever space allows, prefer <strong>direct labels</strong>: place the series name next to the line or bar at the last data point.</p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Guidelines:</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li>With <strong>2–3 series</strong>: always try direct labels.</li>
    <li>With <strong>4–6 series</strong>: use a legend positioned outside the plot area so it does not overlap data.</li>
    <li>With <strong>7+ series</strong>: reconsider the chart design — you likely need to facet, highlight, or aggregate.</li>
    <li>Never rely on colour alone to distinguish categories; also vary line style, marker shape, or use direct labels for accessibility.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Ordering Categories</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Alphabetical order is almost never the right choice for a bar chart. The human eye reads ranking effortlessly when bars are sorted, but must actively search when they are not.</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li><strong>Ranked data</strong> (sales, scores, counts): sort descending so the largest value is first.</li>
    <li><strong>Time data</strong>: always sort chronologically — never alphabetically.</li>
    <li><strong>Ordinal data</strong> (Low / Medium / High): preserve the natural order.</li>
    <li><strong>Geographic data</strong>: group by region or sort by value — not alphabetically by country name.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">In pandas/matplotlib: <code>df.sort_values('value', ascending=False)</code> before plotting, or pass <code>order=</code> to seaborn.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Reducing Clutter — Avoiding Chartjunk</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Edward Tufte coined the term <strong>chartjunk</strong> for visual elements that consume ink without conveying information. Every pixel in a chart should earn its place. Common offenders:</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li><strong>3-D effects</strong> on 2-D data — distort proportions and add nothing.</li>
    <li><strong>Excessive gridlines</strong> — keep only major horizontal gridlines, and make them light grey.</li>
    <li><strong>Heavy border boxes</strong> around the plot area — remove the top and right spines (<code>ax.spines['top'].set_visible(False)</code>).</li>
    <li><strong>Redundant legends</strong> when the chart has only one series.</li>
    <li><strong>Gradient fills</strong> on bars — use flat, single colours.</li>
    <li><strong>Clip-art or background images</strong> behind the data.</li>
    <li><strong>Every bar a different colour</strong> when there is only one variable — use one colour with purposeful highlighting.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><strong>Tufte&rsquo;s data-ink ratio</strong> = (ink used for data) / (total ink). Maximise it.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">White Space and Aspect Ratio</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">White space is not wasted space — it is <strong>breathing room</strong> that guides the eye and reduces fatigue.</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li>Use <code>plt.tight_layout()</code> or <code>fig.subplots_adjust()</code> to prevent labels from being clipped.</li>
    <li>Avoid cramming multiple charts into a tiny figure — prefer larger individual panels.</li>
    <li><strong>Aspect ratio</strong> matters: a wide, shallow chart makes trends look flat; a tall, narrow chart exaggerates slopes. The classic guidance (Cleveland, 1993) is to <strong>bank to 45°</strong> — choose width and height so that the dominant line segments average about 45° from horizontal. For most bar charts, a 4:3 or 16:9 ratio works well.</li>
    <li>Add <strong>padding between bar groups</strong> (<code>width=0.6</code>) so bars do not touch.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Annotation Best Practices</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Annotations are directed attention — they say <em>"look here."</em> Used sparingly, they are powerful. Used indiscriminately, they become noise.</p>
  <ul style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li>Annotate <strong>one to three</strong> key points at most — the peak, an anomaly, or the policy-change date.</li>
    <li>Use a short, plain-language label: <em>"COVID lockdown begins"</em> not <em>"Event X."</em></li>
    <li>Place annotations so they do not overlap data or each other — use <code>ax.annotate()</code> with <code>arrowprops</code> to point at a data point from a clear area.</li>
    <li>Match annotation font size to the chart hierarchy: smaller than the title, similar to or slightly smaller than axis labels.</li>
    <li>Avoid boxing every data value — data labels on all bars become clutter unless precision is essential.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Storytelling with Graphs</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Data visualisation is a communication act. Before you open a code editor, answer three questions:</p>
  <ol style="color: #333; line-height: 1.9; font-size: 1.05em;">
    <li><strong>Who is the audience?</strong> A subject-matter expert can handle more complexity than a general public audience.</li>
    <li><strong>What is the single main message?</strong> If you cannot state it in one sentence, you are not ready to plot.</li>
    <li><strong>What action or decision should the chart enable?</strong> Visualisations exist to help people do something — decide, investigate, or understand.</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Encode the main message visually: highlight the key bar in a contrasting colour while muting all others, draw a reference line at a target value, or use a sequential palette where darker means more. The chart itself should whisper the conclusion before the reader reads a single word.</p>
</div>

## Demonstration: Bad vs Good — Annotated Examples

The following cells show **intentionally poor** visualisations alongside instructor commentary, then **professional corrected versions**. Study the differences carefully.

<div style="background-color: #FFEBEE; border-left: 5px solid #C62828; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #C62828; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Bad Example 1 — Overloaded Pie Chart</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><strong>What the data shows:</strong> Monthly unit sales for a retail store across 12 months.</p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><strong>Why this visualisation fails:</strong></p>
  <ul style="color: #333; line-height: 1.8; font-size: 1.05em;">
    <li>A pie chart with <strong>12 slices</strong> is unreadable — humans cannot accurately judge angles between similar-sized slices.</li>
    <li>There are <strong>no labels</strong> on the slices, forcing the reader to match tiny colour swatches to a legend.</li>
    <li>The <strong>rainbow colour palette</strong> has no semantic meaning and creates visual chaos.</li>
    <li>Pie charts communicate <strong>part-of-whole</strong> relationships, but the real question here is <em>trend over time</em> — a fundamentally different question that a pie chart cannot answer.</li>
    <li>No title, no data source, no units.</li>
  </ul>
</div>

````python
import matplotlib.pyplot as plt
import numpy as np

# --- BAD EXAMPLE 1: Overloaded pie chart ---
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [4200, 3800, 5100, 5600, 6200, 7100,
         7800, 7400, 6300, 5900, 6800, 8500]

# Garish rainbow colours
bad_colors = ['#FF0000', '#FF7F00', '#FFFF00', '#00FF00',
              '#00FFFF', '#0000FF', '#8B00FF', '#FF00FF',
              '#FF1493', '#00FA9A', '#FF4500', '#1E90FF']

fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(sales, colors=bad_colors)  # no labels, no percentages, no title
plt.show()
````

  <h3 style="color: #388E3C; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Corrected Version 1 — Horizontal Bar Chart (Ranked)</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><strong>Improvements made:</strong></p>
  <ul style="color: #333; line-height: 1.8; font-size: 1.05em;">
    <li>Switched to a <strong>horizontal bar chart</strong> — bar lengths are far easier to compare than angles.</li>
    <li>Bars are <strong>sorted by value</strong> (descending) so ranking is immediately visible.</li>
    <li>A <strong>single colour</strong> is used for all bars, with the peak month (December) highlighted in gold to draw the eye to the main finding.</li>
    <li>Clear <strong>title stating the insight</strong>, labelled axes with units, and a data-source note.</li>
    <li>Top and right spines removed; gridlines are light and minimal.</li>

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- CORRECTED EXAMPLE 1: Horizontal bar chart, sorted, minimal ---
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [4200, 3800, 5100, 5600, 6200, 7100,
         7800, 7400, 6300, 5900, 6800, 8500]

df = pd.DataFrame({'month': months, 'sales': sales})
df = df.sort_values('sales', ascending=True)  # ascending for horizontal (largest at top)

# Highlight the peak month
colors = ['#E8B923' if m == 'Dec' else '#4B2E83' for m in df['month']]

fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.barh(df['month'], df['sales'], color=colors, height=0.6)

# Minimal value labels only on the highlighted bar
for bar, month in zip(bars, df['month']):
    if month == 'Dec':
        ax.text(bar.get_width() + 80, bar.get_y() + bar.get_height() / 2,
                f"{int(bar.get_width()):,} units",
                va='center', fontsize=10, color='#E8B923', fontweight='bold')

# Titles and labels
ax.set_title('December Led Annual Sales with 8,500 Units Sold',
             fontsize=13, color='#4B2E83', fontweight='bold', pad=14)
ax.set_xlabel('Units Sold', fontsize=11, color='#333')
ax.set_ylabel('Month', fontsize=11, color='#333')

# Remove chartjunk
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.grid(True, color='#DDDDDD', linewidth=0.8)
ax.set_axisbelow(True)

fig.text(0.99, -0.01, 'Source: Retail store internal data, 2023',
         ha='right', fontsize=8, color='#888')
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Accessibility Matters</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Approximately <strong>8% of males and 0.5% of females</strong> of Northern European descent have some form of color vision deficiency (CVD). In a classroom of 30 students, there is a very good chance that 2&ndash;3 people cannot distinguish your red from your green. Globally, that is roughly 300 million people. When a chart relies solely on color to encode information, those viewers receive no information at all from that encoding channel. Accessible visualization is not a "nice to have" &mdash; it is a basic requirement for communicating with your full audience. The Web Content Accessibility Guidelines (WCAG 2.1) require a contrast ratio of at least <strong>4.5:1</strong> for normal text and <strong>3:1</strong> for large text and graphical elements. These same principles apply to data graphics.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Common Forms of Color Blindness</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Color blindness is not "seeing in black and white" &mdash; it is a shift or absence of sensitivity in one cone type. The three main types are:
  </p>
  <ul style="font-size: 1.05em; color: #555; line-height: 2;">
    <li><strong>Deuteranopia / Deuteranomaly</strong> &mdash; reduced or absent green cone response (most common, ~5% of males). Red and green appear as shades of yellow-brown.</li>
    <li><strong>Protanopia / Protanomaly</strong> &mdash; reduced or absent red cone response (~1% of males). Reds appear dark, greens look yellow-green.</li>
    <li><strong>Tritanopia / Tritanomaly</strong> &mdash; reduced or absent blue cone response (very rare, affects both sexes equally). Blues appear green, yellows appear pink.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>The practical implication:</strong> red&ndash;green distinctions are the most commonly broken encoding pair in data visualization. A pie chart with a red slice and a green slice is completely ambiguous to ~8% of male viewers. Deuteranopia and protanopia together account for the vast majority of CVD cases.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Grayscale Readability and Contrast</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A simple but powerful test is to convert your chart to grayscale (or print it on a black-and-white printer) and ask: <em>can the viewer still read every data series?</em> If two categories collapse to the same gray tone, color is doing all the work and your chart fails the test. The fix involves adding redundant encodings:
  </p>
  <ul style="font-size: 1.05em; color: #555; line-height: 2;">
    <li><strong>Markers</strong> on line charts (circle, square, triangle, diamond&hellip;)</li>
    <li><strong>Line styles</strong> (solid, dashed, dotted, dash-dot)</li>
    <li><strong>Hatching / fill patterns</strong> on bars and areas</li>
    <li><strong>Direct labels</strong> placed next to the last data point or bar, eliminating the need for a legend entirely</li>
    <li><strong>Shape</strong> in scatter plots instead of relying on color alone</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    WCAG 2.1 guideline 1.4.1 states: "Color is not used as the only visual means of conveying information, indicating an action, prompting a response, or distinguishing a visual element."
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Avoiding Rainbow (Jet) Colormaps for Numeric Data</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <code>jet</code> (rainbow) colormap was the matplotlib default for many years and is still the default in MATLAB. It is <strong>perceptually non-uniform</strong>: equal steps in data value do not correspond to equal steps in perceived color. The green and cyan bands appear much lighter than the red and blue ends, creating false visual boundaries that do not exist in the data. In addition, jet fails for colorblind viewers and converts to a confusing non-monotonic grayscale.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The modern replacements &mdash; <strong>viridis, cividis, plasma, magma, inferno</strong> &mdash; were designed to be:
  </p>
  <ul style="font-size: 1.05em; color: #555; line-height: 2;">
    <li>Perceptually uniform (equal data steps = equal perceived steps)</li>
    <li>Monotonically increasing in lightness (converts cleanly to grayscale)</li>
    <li>Distinguishable for the most common forms of CVD</li>
    <li><code>cividis</code> specifically is optimized for both deuteranopia and protanopia</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">ColorBrewer and Palette Principles</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>ColorBrewer</strong> (colorbrewer2.org), developed by cartographer Cynthia Brewer, provides three families of palettes:
  </p>
  <ul style="font-size: 1.05em; color: #555; line-height: 2;">
    <li><strong>Sequential</strong> &mdash; for ordered data going from low to high (e.g., population density). Use one hue that varies in lightness.</li>
    <li><strong>Diverging</strong> &mdash; for data with a meaningful midpoint (e.g., temperature anomaly relative to 1990 baseline). Two hues meet at a neutral center.</li>
    <li><strong>Qualitative</strong> &mdash; for categorical data with no inherent order. Maximally distinct hues at similar lightness levels.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Many ColorBrewer palettes are labeled as "colorblind safe" on the website. In Python, they are available via <code>matplotlib.cm</code> (e.g., <code>RdYlBu</code>, <code>BuGn</code>) and <code>seaborn</code> (e.g., <code>sns.color_palette('colorblind')</code>, <code>sns.color_palette('Set2')</code>).
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Quick palette resources:</strong>
    <a href="https://colorhunt.co" style="color: #4B2E83;">colorhunt.co</a> and
    <a href="https://coolors.co" style="color: #4B2E83;">coolors.co</a> let you browse and copy hex codes for custom palettes.
    Both sites provide CVD simulation previews. Once you have hex codes, passing them directly to matplotlib or seaborn is straightforward.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Inclusive Dashboard Design Principles</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    When assembling a multi-panel dashboard, apply these principles consistently:
  </p>
  <ol style="font-size: 1.05em; color: #555; line-height: 2;">
    <li><strong>Never encode information in color alone.</strong> Pair color with shape, pattern, or direct annotation.</li>
    <li><strong>Use the same palette across all panels</strong> so that a color means the same category everywhere.</li>
    <li><strong>Avoid more than 6&ndash;8 categorical colors.</strong> Beyond that, use facets or small multiples.</li>
    <li><strong>Check contrast ratios</strong> between text annotations and their backgrounds using a contrast checker.</li>
    <li><strong>Prefer direct labels over legends</strong> wherever space allows &mdash; legends add a lookup step that slows comprehension.</li>
    <li><strong>Test with a grayscale print</strong> before publishing.</li>
    <li><strong>Use CVD simulation tools</strong> (e.g., the <code>colorspacious</code> Python library, Coblis, or the macOS Display Accessibility filter) to preview your charts as a deuteranopic viewer sees them.</li>
  </ol>

<div style="border-left: 5px solid #C62828; background-color: #FFF8F8; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #C62828; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Bad Example 2 — Rainbow (Jet) Heatmap</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <code>jet</code> colormap appears vibrant but introduces false visual boundaries: the bright yellow-green band makes mid-range values look discontinuously different from adjacent values. It is perceptually non-uniform, fails for colorblind viewers, and converts to a non-monotonic grayscale. Here we apply it to a heatmap of average monthly temperature (°C) across several cities.
  </p>
</div>

````python
# ---- BAD EXAMPLE 2: rainbow/jet heatmap ----

cities = ['Seattle', 'Phoenix', 'Chicago', 'Miami', 'Denver', 'Boston']
month_labels = ['Jan','Feb','Mar','Apr','May','Jun',
                'Jul','Aug','Sep','Oct','Nov','Dec']

# Realistic average monthly temperatures (°C)
temp_data = np.array([
    [4,  5,  7, 10, 14, 17, 20, 20, 17, 12,  7,  4],   # Seattle
    [12, 14, 18, 23, 28, 34, 37, 36, 32, 25, 17, 11],   # Phoenix
    [-4, -2,  4, 11, 17, 22, 25, 24, 19, 12,  4, -2],   # Chicago
    [20, 20, 22, 24, 26, 28, 30, 30, 29, 27, 24, 21],   # Miami
    [-1,  1,  5, 10, 15, 20, 24, 23, 18, 11,  4, -1],   # Denver
    [-2,  0,  4,  9, 15, 20, 23, 22, 18, 12,  6,  0],   # Boston
])

fig, ax = plt.subplots(figsize=(11, 4))
im = ax.imshow(temp_data, cmap='jet', aspect='auto')
plt.colorbar(im, ax=ax, label='Avg Temperature (°C)')
ax.set_xticks(range(12))
ax.set_xticklabels(month_labels)
ax.set_yticks(range(len(cities)))
ax.set_yticklabels(cities)
ax.set_title('Average Monthly Temperature by City', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.suptitle('BAD: jet/rainbow colormap — perceptually non-uniform, CVD-unsafe',
             y=1.02, color='#C62828', fontsize=11, style='italic')
plt.show()
````

<div style="border-left: 5px solid #388E3C; background-color: #F4FBF4; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #388E3C; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Corrected Example 2 — Viridis Heatmap with Value Annotations</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Replacing <code>jet</code> with <code>viridis</code> (or <code>cividis</code> for maximum CVD safety) gives a perceptually uniform gradient. We also annotate each cell with the numeric value, so the reader does not need to mentally decode color into number. Notice how the gradient now smoothly represents the true ordering of temperatures without any false boundaries.
  </p>
</div>

````python
# ---- CORRECTED EXAMPLE 2: viridis heatmap + cell annotations ----

fig, axes = plt.subplots(1, 2, figsize=(16, 4))

for ax, cmap_name, title_suffix in zip(
        axes,
        ['viridis', 'cividis'],
        ['viridis  (perceptually uniform)', 'cividis  (optimized for deuteranopia/protanopia)']):

    im = ax.imshow(temp_data, cmap=cmap_name, aspect='auto',
                   vmin=temp_data.min(), vmax=temp_data.max())
    plt.colorbar(im, ax=ax, label='Avg Temp (°C)')
    ax.set_xticks(range(12))
    ax.set_xticklabels(month_labels, fontsize=9)
    ax.set_yticks(range(len(cities)))
    ax.set_yticklabels(cities)
    ax.set_title(title_suffix, fontsize=11, fontweight='bold')

    # Annotate each cell with the value
    for r in range(temp_data.shape[0]):
        for c in range(temp_data.shape[1]):
            val = temp_data[r, c]
            text_color = 'white' if val < 15 else 'black'
            ax.text(c, r, f'{val}', ha='center', va='center',
                    fontsize=7.5, color=text_color, fontweight='bold')

plt.suptitle('GOOD: Perceptually uniform colormaps + cell annotations',
             y=1.03, color='#388E3C', fontsize=11, style='italic')
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Demo — Importing a Custom Hex Palette from coolors.co into Python</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    coolors.co and colorhunt.co generate palettes you can copy as a list of hex codes. The demo below shows how to paste those hex codes into Python, register them as a seaborn palette, and preview the result. We also simulate a rough deuteranopia view using a luminance-based grayscale transform.
  </p>
</div>

````python
# ---- DEMO: Import hex palette from coolors.co, preview + grayscale simulation ----

# Paste your hex codes from coolors.co here (this is a CVD-safe 5-color palette)
coolors_palette = ['#264653', '#2A9D8F', '#E9C46A', '#F4A261', '#E76F51']

# ---- 1. Register as a seaborn palette and preview ----
sns.set_palette(coolors_palette)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Panel A: color swatches
ax = axes[0]
for i, hex_color in enumerate(coolors_palette):
    ax.barh(i, 1, color=hex_color, height=0.8)
    ax.text(0.5, i, hex_color, ha='center', va='center',
            color='white' if i in [0, 1] else '#333',
            fontweight='bold', fontsize=11)
ax.set_xlim(0, 1)
ax.set_yticks(range(len(coolors_palette)))
ax.set_yticklabels([f'Color {i+1}' for i in range(len(coolors_palette))])
ax.set_title('Custom Palette (coolors.co)', fontweight='bold')
ax.axis('off')
ax.set_frame_on(False)

# Panel B: bar chart using the custom palette
ax = axes[1]
categories = ['North', 'South', 'East', 'West', 'Central']
values = [34, 51, 42, 29, 46]
bars = ax.bar(categories, values, color=coolors_palette, width=0.6)
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            str(val), ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_title('Regional Sales (custom palette)', fontweight='bold')
ax.set_ylabel('Sales ($M)')
ax.set_ylim(0, 60)

# Panel C: grayscale simulation
# Convert hex to luminance using WCAG relative luminance formula
def hex_to_gray(hex_color):
    r, g, b = int(hex_color[1:3], 16)/255, int(hex_color[3:5], 16)/255, int(hex_color[5:7], 16)/255
    # Linearize
    def lin(c): return c/12.92 if c <= 0.03928 else ((c+0.055)/1.055)**2.4
    L = 0.2126*lin(r) + 0.7152*lin(g) + 0.0722*lin(b)
    return (L, L, L)  # RGB gray

gray_palette = [hex_to_gray(c) for c in coolors_palette]

ax = axes[2]
bars_g = ax.bar(categories, values, color=gray_palette, width=0.6,
                edgecolor='#555', linewidth=0.8)
for bar, val in zip(bars_g, values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            str(val), ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_title('Grayscale simulation', fontweight='bold')
ax.set_ylabel('Sales ($M)')
ax.set_ylim(0, 60)

plt.suptitle('Importing and Testing a Custom Palette from coolors.co',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('Grayscale luminance values:', [f'{(hex_to_gray(c)[0]):.3f}' for c in coolors_palette])
print('A well-separated palette should have at least 0.10 difference between adjacent luminances.')
````

    This side-by-side scatter plot compares an inaccessible version (only color distinguishes groups) against an accessible version (color + shape + direct annotation). The accessible version works even when printed in black and white.

````python
# ---- DEMO: Accessible vs inaccessible scatter plots ----

# Simulated student data: study hours vs exam score, by major
np.random.seed(7)
n = 50
majors = ['Engineering', 'Business', 'Arts', 'Sciences']
colors_bad  = ['#D32F2F', '#388E3C', '#1565C0', '#F57F17']  # red/green problem
colors_good = ['#1976D2', '#F57C00', '#7B1FA2', '#388E3C']   # CVD-safe
markers     = ['o', 's', '^', 'D']

study_hours = {m: np.random.normal(5 + i, 1.5, n) for i, m in enumerate(majors)}
exam_scores = {m: 50 + study_hours[m] * 4 + np.random.normal(0, 5, n)
               for m in majors}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ---- Left: BAD (color only, same marker) ----
ax = axes[0]
for i, major in enumerate(majors):
    ax.scatter(study_hours[major], exam_scores[major],
               color=colors_bad[i], s=50, alpha=0.75, marker='o', label=major)
ax.set_title('BAD: Color only (red/green present)', color='#C62828',
             fontweight='bold', fontsize=12)
ax.set_xlabel('Study Hours per Week')
ax.set_ylabel('Exam Score')
ax.legend(title='Major', framealpha=0.9)
ax.set_xlim(0, 12)

# ---- Right: GOOD (color + shape + direct labels) ----
ax = axes[1]
for i, major in enumerate(majors):
    ax.scatter(study_hours[major], exam_scores[major],
               color=colors_good[i], s=55, alpha=0.75,
               marker=markers[i], label=major)
    # Annotate cluster centroid
    cx = np.mean(study_hours[major])
    cy = np.mean(exam_scores[major])
    ax.text(cx + 0.15, cy, major, color=colors_good[i],
            fontsize=8.5, fontweight='bold', va='center')

ax.set_title('GOOD: Color + shape + direct labels', color='#388E3C',
             fontweight='bold', fontsize=12)
ax.set_xlabel('Study Hours per Week')
ax.set_ylabel('Exam Score')
ax.legend(title='Major', framealpha=0.9)
ax.set_xlim(0, 12)

plt.suptitle('Accessible vs Inaccessible Scatter Plots',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Hue, Saturation, and Brightness</h2>
</div>

Every color has three independent dimensions. Understanding them gives you precise control over any palette.

| Dimension | Definition | Effect on charts |
|:----------|:-----------|:-----------------|
| **Hue** | The *identity* of the color — red, blue, green, etc. Defined by wavelength on the color wheel. | Use distinct hues to encode **categorical** distinctions (product A vs. product B). |
| **Saturation** | The *intensity* or purity of the hue — 0 % is pure gray, 100 % is the most vivid form. | High saturation draws the eye. Reserve it for the **accent** (10 %) role. Mute everything else. |
| **Brightness (Value)** | The *lightness* or darkness — 0 % is black, 100 % is white. | Light backgrounds reduce eye strain; dark backgrounds suit presentation screens but hurt prints. |

> **Practical rule:** For most business charts, keep background saturation near 0 (neutral gray or white), secondary elements at low-to-medium saturation, and your single accent element at full or near-full saturation.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Warm vs. Cool Colors</h2>
</div>

The color wheel is commonly split into **warm** (red, orange, yellow) and **cool** (blue, green, purple) families. Each carries inherited cultural and psychological associations that your audience will apply — consciously or not — to your data.

| Family | Colors | Typical perception | Common data application |
|:-------|:-------|:-------------------|:------------------------|
| **Warm** | Red, Orange, Yellow | Urgency, danger, energy, heat, positivity (yellow) | Alerts, anomalies, high-risk KPIs, profit/growth |
| **Cool** | Blue, Green, Purple | Calm, trust, stability, coldness | Neutral comparisons, brand colors, financial dashboards |

**Why it matters for visualization:**
- In a dashboard showing **budget overruns**, mapping red to over-budget and blue to on-target leverages instinctive associations — no legend needed.
- If you invert this (blue for danger, red for safe), the audience fights the chart.
- **Diverging palettes** often run from a warm extreme (red) through a neutral center (white or gray) to a cool extreme (blue), which aligns naturally with positive/negative or above-average/below-average values.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Categorical, Sequential, and Diverging Palettes</h2>
</div>

Choosing the wrong palette type is one of the most common color mistakes in data visualization. Each type is designed for a specific kind of data.

### Categorical Palette
- **Use when:** Your variable has distinct, unordered groups (product lines, regions, departments).
- **Design:** Maximally distinct hues at similar brightness/saturation so no group looks more important.
- **Limit:** 5–7 categories maximum before colors become too similar to distinguish.
- **Example:** `[#4B2E83, #E8B923, #1976D2, #388E3C, #F57C00]`

### Sequential Palette
- **Use when:** Your variable has a natural order from low to high (revenue, temperature, count).
- **Design:** One hue progressively darkens (or brightens). Darker typically means higher.
- **Mistake to avoid:** Using a rainbow (jet) colormap — it introduces false boundaries where colors change abruptly.
- **Example:** Light lavender → deep purple for a revenue heatmap.

### Diverging Palette
- **Use when:** Your variable has a meaningful midpoint (zero, average, target) and deviates in two directions.
- **Design:** Two contrasting hues fade toward a neutral center.
- **Example:** Red → white → blue for a profit-vs-loss chart, or a correlation heatmap ranging from -1 to +1.

> **Quick check:** Ask yourself: *Does the order of the data values matter?* If no — categorical. If yes — *Is there a meaningful midpoint?* If yes — diverging. If no — sequential.

Audiences do not read color as neutral information — they *feel* it. Colors trigger associations built from culture, nature, and experience. Ignoring this is leaving a powerful communication tool unused.

| Color | Common association | Data application |
|:------|:-------------------|:-----------------|
| **Red** `#C62828` | Danger, loss, urgency, stop | Budget overruns, declines, alerts, negative values |
| **Green** `#388E3C` | Growth, safety, go, profit | Positive KPIs, on-target metrics, increases |
| **Blue** `#1976D2` | Trust, calm, technology, neutrality | Default comparisons, background categories, corporate reports |

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Visual Hierarchy and Emphasis Through Accent Color</h2>
</div>

**Visual hierarchy** is the principle that not everything in a chart deserves equal attention. Your job as a designer is to guide the viewer's eye to the most important element first.

Color is the fastest tool for creating hierarchy:

| Role | Color treatment | Viewer response |
|:-----|:----------------|:----------------|
| **Accent (10%)** | High saturation, distinctive hue | Eye goes here first — this is the key insight |
| **Secondary (20%)** | Medium saturation, muted hue | Supports context without competing |
| **Base/Neutral (70%)** | Low saturation, gray or off-white | Provides structure without distraction |

**Emphasis through accent color** means reserving one vivid color exclusively for the element you want the audience to notice — the best-performing product, the month revenue crossed the target, the anomalous data point.

> **The accent rule:** If everything is highlighted, nothing is. One accent color. Use it once per chart, or at most for one category across all charts in a report.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Neutral vs. Saturated Backgrounds</h2>
</div>

The background is the largest color area in any chart. It sets the baseline against which every data element is perceived.

**Neutral backgrounds (white, light gray, off-white):**
- Standard for printed reports and most business dashboards.
- Data elements pop cleanly against them.
- Gridlines and axes remain visible without competing.
- Recommended: `#FFFFFF`, `#F5F5F5`, `#FAFAFA`.

**Saturated / dark backgrounds:**
- Work well on presentation screens (conferences, digital dashboards viewed from a distance).
- Require higher-contrast text and lighter data elements.
- Can create a dramatic, premium feel — but exhaust eyes in long-form reading.
- Use sparingly: typically only for hero slides or executive scorecards.

**The rule:** Match your background choice to the **medium** and **context** of consumption, not personal aesthetic preference.

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Consistency Across Multi-Chart Dashboards</h2>
</div>

A dashboard is a collection of related charts. Inconsistent color use across panels creates cognitive friction — the viewer must re-learn what each color means with every chart they look at.

**Rules for multi-chart consistency:**

1. **Same category = same color.** If "Q4 Revenue" is `#4B2E83` in the bar chart, it must be `#4B2E83` in the line chart and the scatter plot.
2. **Same role = same treatment.** If gridlines are `#DDDDDD` and 30 % opacity in one chart, use those exact values everywhere.
3. **One accent, used consistently.** If your accent color flags the KPI of interest, apply it to the same category or value across all panels — never repurpose it for a different meaning in a different chart.
4. **Consistent background.** All panels on a dashboard should share the same figure and axes background colors.
5. **Define a palette dictionary at the top of your code** and import it into every chart function — never hard-code a hex string twice.

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Rule Explained</h2>
</div>

The 70-20-10 rule borrows from interior design and brand identity to provide a simple, memorable framework for allocating color in any visual composition.

| Proportion | Role | Typical colors | Typical chart elements |
|:----------:|:-----|:---------------|:-----------------------|
| **70%** | **Dominant / Base** | Neutral: white, light gray, off-white | Figure background, axes area, gridlines, un-highlighted bars/points |
| **20%** | **Secondary / Supporting** | Muted brand color or medium-saturation hue | Secondary category bars, axis labels, tick marks, legend text, supporting lines |
| **10%** | **Accent / Highlight** | Vivid, high-saturation color | The KPI bar, the anomaly point, the record-breaking month, the key insight |

### Why it works

Human vision evolved to detect contrast. When 90 % of a chart is quiet (neutral or muted), the 10 % accent creates an automatic, effortless focal point. The viewer does not need to read a caption — their eye is already on the insight.

Without the rule, a chart where every element screams for attention forces the viewer to do all the interpretive work. With it, the chart does the work for them.

</div>

Here is how the three zones map to the concrete elements of a typical multi-panel KPI dashboard:

**70% — Neutral base:**
- Figure background: `#FAFAFA` (near-white)
- Axes background: `white` or very light gray
- Gridlines: `#DDDDDD`, opacity 40 %
- Non-highlighted bars, lines, and scatter points: `#CCCCCC` (neutral gray)
- Axis tick labels, axis titles: `#555555`

**20% — Supporting color:**
- Secondary categories (e.g., all product lines except the top performer): `#B7A6D5` (muted lavender)
- Supporting annotation text
- Legend markers
- Chart titles: `#4B2E83` (brand color, but restrained by area)

**10% — Accent:**
- The top-performing bar: `#C62828` (bold red) or `#E8B923` (gold)
- The anomalous data point: `#F57C00` (orange)
- The month revenue crossed a target line: `#4B2E83` (primary brand, full saturation)
- KPI tiles that are off-track: `#C62828`

> **Remember:** The accent does its job because everything else is quiet. Resist the temptation to use more than one accent per chart.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)

# ── Shared palette ─────────────────────────────────────────────────────────────
PALETTE = {
    'primary':   '#4B2E83',
    'gold':      '#E8B923',
    'dark':      '#32006E',
    'lavender':  '#B7A6D5',
    'tan':       '#85714D',
    'red':       '#C62828',
    'blue':      '#1976D2',
    'orange':    '#F57C00',
    'green':     '#388E3C',
    'purple':    '#7B1FA2',
    'neutral':   '#CCCCCC',
    'neutral2':  '#AAAAAA',
    'bg':        '#FAFAFA',
}

# ── Realistic business KPI data ────────────────────────────────────────────────
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

revenue      = [4.1, 3.8, 4.5, 5.2, 5.8, 6.1, 5.9, 6.3, 7.1, 7.4, 8.2, 9.0]
expenses     = [3.2, 3.1, 3.4, 3.6, 3.8, 4.0, 3.9, 4.1, 4.3, 4.5, 4.8, 5.1]
headcount    = [142, 145, 148, 150, 153, 155, 154, 157, 160, 162, 165, 170]
satisfaction = [7.2, 7.0, 7.4, 7.5, 7.6, 7.3, 7.8, 7.9, 8.1, 8.0, 8.3, 8.5]

departments  = ['Sales', 'Marketing', 'Engineering', 'Operations', 'Finance', 'HR']
dept_revenue = [42.3, 18.7, 8.1, 12.4, 9.9, 5.2]

print('Setup complete ✓')
````

### Bad Example 1 — The Competing-Color Dashboard

The chart below shows a realistic 2×2 business dashboard. Every bar is a different bright color, every panel uses a different scheme, and there is no visual hierarchy. The result is visually exhausting — the eye has nowhere to rest, and there is no obvious message.

````python
# ── BAD EXAMPLE 1: Competing saturated colors, no visual hierarchy ──────────────
bad_colors_bars   = ['#E53935','#8E24AA','#1E88E5','#43A047','#FB8C00','#00ACC1']
bad_colors_line1  = '#FF1744'
bad_colors_line2  = '#00E676'
bad_colors_hc     = ['#FFEB3B','#FF5722','#9C27B0','#00BCD4',
                      '#8BC34A','#FF9800','#607D8B','#F44336',
                      '#3F51B5','#009688','#CDDC39','#795548']
bad_colors_sat    = ['#E91E63','#2196F3','#4CAF50','#FF5722',
                      '#9C27B0','#FFC107','#00BCD4','#8BC34A',
                      '#FF9800','#F44336','#673AB7','#03A9F4']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.patch.set_facecolor('#FFFFFF')
fig.suptitle('Q4 Business KPI Dashboard', fontsize=18, fontweight='bold', color='#000000')

# Panel 1: Revenue by department — every bar a different saturated color
ax1 = axes[0, 0]
ax1.bar(departments, dept_revenue, color=bad_colors_bars)
ax1.set_title('Revenue by Department ($M)', fontsize=13, color='#E53935')
ax1.set_facecolor('#FFFDE7')

# Panel 2: Monthly revenue vs expenses — competing neon lines
ax2 = axes[0, 1]
ax2.plot(months, revenue,  color=bad_colors_line1, linewidth=3, marker='D', label='Revenue')
ax2.plot(months, expenses, color=bad_colors_line2, linewidth=3, marker='*', label='Expenses')
ax2.set_title('Revenue vs Expenses ($M)', fontsize=13, color='#1E88E5')
ax2.legend()
ax2.set_facecolor('#E8F5E9')

# Panel 3: Headcount — every bar a random color
ax3 = axes[1, 0]
ax3.bar(months, headcount, color=bad_colors_hc)
ax3.set_title('Headcount by Month', fontsize=13, color='#9C27B0')
ax3.set_facecolor('#FCE4EC')

# Panel 4: Satisfaction scores — random scatter colors
ax4 = axes[1, 1]
ax4.scatter(range(12), satisfaction, c=bad_colors_sat, s=150, zorder=3)
ax4.plot(range(12), satisfaction, color='#000000', linewidth=1, zorder=2)
ax4.set_title('Employee Satisfaction', fontsize=13, color='#FF5722')
ax4.set_facecolor('#E3F2FD')

for ax in axes.flat:
    ax.grid(True, alpha=0.5)
    ax.tick_params(axis='x', rotation=45)

fig.text(0.5, -0.02,
    '⚠ BAD: Every element uses a competing saturated color. '
    'No hierarchy. No focal point. The viewer does not know where to look.',
    ha='center', fontsize=11, color='#C62828', style='italic')

plt.tight_layout()
plt.show()
````

**What is wrong with this chart?**

- Every bar and every data point is a maximally saturated, distinct color — yet the colors encode *no information*. The blue bar in the department chart is not more important than the red one; they are equally loud.
- Each panel uses a completely different background color and color scheme, so the four charts feel like they came from four different designers.
- There is no accent element — nothing that says "this is the number you need to care about."
- The viewer's cognitive load spikes immediately. Processing the chart takes effort rather than being effortless.

Now let us correct it using the 70-20-10 rule.

````python
# ── CORRECTED EXAMPLE 1: 70-20-10 applied to the same dashboard ────────────────
#
# 70% neutral: white/near-white backgrounds, gray bars/points
# 20% secondary: muted lavender for supporting categories / lines
# 10% accent: UW gold #E8B923 for the KPI that matters most

BASE    = PALETTE['neutral']     # 70% — non-highlighted bars, points
SUPPORT = PALETTE['lavender']    # 20% — secondary line, supporting labels
ACCENT  = PALETTE['gold']        # 10% — the KPI we want to call out
TITLE_C = PALETTE['primary']

# Top-performing department: Sales (index 0)
dept_colors = [ACCENT if i == 0 else BASE for i in range(len(departments))]

# Month with highest revenue: December (index 11)
hc_colors   = [ACCENT if i == 11 else BASE for i in range(12)]
sat_colors  = [ACCENT if s == max(satisfaction) else BASE for s in satisfaction]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.patch.set_facecolor(PALETTE['bg'])
fig.suptitle('Q4 Business KPI Dashboard', fontsize=18, fontweight='bold', color=TITLE_C)

# Panel 1: Revenue by department — Sales highlighted as top earner
ax1 = axes[0, 0]
ax1.set_facecolor('white')
bars = ax1.bar(departments, dept_revenue, color=dept_colors, edgecolor='white', linewidth=0.8)
ax1.bar_label(bars, fmt='$%.1fM', fontsize=9, padding=3, color=TITLE_C)
ax1.set_title('Revenue by Department ($M)\n', fontsize=13, fontweight='bold', color=TITLE_C)
ax1.set_ylabel('Revenue ($M)', fontsize=11)
ax1.tick_params(axis='x', rotation=30)
ax1.annotate('▲ Top earner', xy=(0, dept_revenue[0]), xytext=(0.3, dept_revenue[0] + 1.5),
             color=ACCENT, fontsize=9, fontweight='bold')
ax1.grid(True, axis='y', alpha=0.35, color='#DDDDDD')
ax1.spines[['top', 'right']].set_visible(False)

# Panel 2: Revenue vs expenses — revenue in primary, expenses in muted support
ax2 = axes[0, 1]
ax2.set_facecolor('white')
ax2.plot(months, expenses, color=SUPPORT, linewidth=2, marker='o', markersize=5,
         label='Expenses', zorder=2)
ax2.plot(months, revenue,  color=TITLE_C,  linewidth=2.5, marker='o', markersize=6,
         label='Revenue', zorder=3)
# Accent: highlight December (record month)
ax2.scatter([11], [revenue[11]], color=ACCENT, s=120, zorder=5, label='Record month')
ax2.axhline(y=7.0, color=PALETTE['red'], linewidth=1, linestyle='--', alpha=0.6, label='Target $7M')
ax2.set_title('Revenue vs Expenses ($M)\n', fontsize=13, fontweight='bold', color=TITLE_C)
ax2.set_ylabel('Amount ($M)', fontsize=11)
ax2.tick_params(axis='x', rotation=30)
ax2.legend(fontsize=9, frameon=False)
ax2.grid(True, alpha=0.35, color='#DDDDDD')
ax2.spines[['top', 'right']].set_visible(False)

# Panel 3: Headcount — December highlighted
ax3 = axes[1, 0]
ax3.set_facecolor('white')
ax3.bar(months, headcount, color=hc_colors, edgecolor='white', linewidth=0.8)
ax3.set_title('Headcount by Month\n', fontsize=13, fontweight='bold', color=TITLE_C)
ax3.set_ylabel('Headcount', fontsize=11)
ax3.tick_params(axis='x', rotation=30)
ax3.annotate('▲ Year-end peak', xy=(11, headcount[11]),
             xytext=(8.5, headcount[11] + 1.5), color=ACCENT, fontsize=9, fontweight='bold')
ax3.grid(True, axis='y', alpha=0.35, color='#DDDDDD')
ax3.spines[['top', 'right']].set_visible(False)

# Panel 4: Satisfaction — best month highlighted
ax4 = axes[1, 1]
ax4.set_facecolor('white')
ax4.plot(range(12), satisfaction, color=SUPPORT, linewidth=2, zorder=2)
ax4.scatter(range(12), satisfaction, c=sat_colors, s=80, zorder=3, edgecolors='white', linewidth=0.5)
ax4.set_title('Employee Satisfaction (1–10)\n', fontsize=13, fontweight='bold', color=TITLE_C)
ax4.set_ylabel('Score', fontsize=11)
ax4.set_xticks(range(12))
ax4.set_xticklabels(months, rotation=30, fontsize=9)
peak_i = satisfaction.index(max(satisfaction))
ax4.annotate(f'Peak {max(satisfaction)}', xy=(peak_i, max(satisfaction)),
             xytext=(peak_i - 2.5, max(satisfaction) - 0.3),
             color=ACCENT, fontsize=9, fontweight='bold',
             arrowprops=dict(arrowstyle='->', color=ACCENT, lw=1.2))
ax4.grid(True, alpha=0.35, color='#DDDDDD')
ax4.spines[['top', 'right']].set_visible(False)

fig.text(0.5, -0.02,
    '✓ CORRECTED: 70% neutral (gray/white), 20% muted lavender/brand, '
    '10% gold accent on the KPI that matters. One focal point per panel.',
    ha='center', fontsize=11, color=PALETTE['green'], style='italic')

plt.tight_layout()
plt.show()
````

**What changed — and why it works now:**

- The background is a consistent near-white `#FAFAFA` across all four panels, providing a quiet base (70%).
- Non-highlighted bars and scatter points are neutral gray `#CCCCCC` — they provide shape and scale without competing for attention.
- Supporting lines and secondary labels use muted lavender `#B7A6D5` (20%).
- The gold accent `#E8B923` appears exactly once per panel, marking the single insight the viewer should leave with: the top revenue department, the record revenue month, the peak headcount month, and the best satisfaction score.
- All panels use the same background, the same gridline style, and the same accent color — the dashboard reads as one cohesive piece.

### Bad Example 2 — The Uniform-Color Scatter Plot

The chart below shows monthly sales calls vs. closed deals for a sales team. All 80 data points are rendered in bright red. There is no hierarchy — the one underperforming rep who is dragging down team numbers is completely invisible.

````python
# ── Sales team data ──────────────────────────────────────────────────────────────
np.random.seed(7)
n_reps   = 79
calls    = np.random.normal(120, 20, n_reps).clip(60, 200).astype(int)
deals    = (calls * np.random.uniform(0.18, 0.35, n_reps)).astype(int)

# One deliberate outlier: very high calls, very low conversion
outlier_calls = np.array([195])
outlier_deals = np.array([8])

all_calls = np.concatenate([calls, outlier_calls])
all_deals = np.concatenate([deals, outlier_deals])

# ── BAD: everything in bright red, no hierarchy ──────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(all_calls, all_deals, color='#FF1744', s=80, alpha=0.8, edgecolors='white', linewidth=0.5)

ax.set_title('Sales Calls vs. Closed Deals — Rep Performance', fontsize=16,
             fontweight='bold', color='#000000')
ax.set_xlabel('Calls Made (Monthly)', fontsize=13)
ax.set_ylabel('Deals Closed (Monthly)', fontsize=13)
ax.grid(True, alpha=0.4)

fig.text(0.1, -0.03,
    '⚠ BAD: All 80 reps rendered in identical bright red. '
    'The underperforming rep with high calls but only 8 closed deals is invisible.',
    fontsize=10, color='#C62828', style='italic')

plt.tight_layout()
plt.show()
````

````python
# ── CORRECTED: 70-20-10 — neutral points, accent on the outlier ─────────────────
#
# 70%  neutral gray for the 79 typical reps
# 20%  muted brand color for the trend line
# 10%  bold orange accent for the single outlier that management needs to act on

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('white')
fig.patch.set_facecolor(PALETTE['bg'])

# 70% neutral — all typical reps
ax.scatter(calls, deals,
           color=PALETTE['neutral'], s=70, alpha=0.75,
           edgecolors='white', linewidth=0.5, label='Sales reps (n=79)', zorder=2)

# 20% supporting — trend line
m, b = np.polyfit(calls, deals, 1)
x_line = np.linspace(calls.min(), calls.max(), 100)
ax.plot(x_line, m * x_line + b,
        color=PALETTE['lavender'], linewidth=2, linestyle='--',
        label='Expected trend', zorder=3)

# 10% accent — the outlier
ax.scatter(outlier_calls, outlier_deals,
           color=PALETTE['orange'], s=160, edgecolors=PALETTE['dark'],
           linewidth=1.5, zorder=5, label='Underperforming rep')
ax.annotate('195 calls → only 8 deals\n⚠ Needs coaching',
            xy=(outlier_calls[0], outlier_deals[0]),
            xytext=(outlier_calls[0] - 60, outlier_deals[0] + 7),
            fontsize=10, color=PALETTE['orange'], fontweight='bold',
            arrowprops=dict(arrowstyle='->', color=PALETTE['orange'], lw=1.5))

ax.set_title('Sales Calls vs. Closed Deals — Rep Performance',
             fontsize=16, fontweight='bold', color=PALETTE['primary'])
ax.set_xlabel('Calls Made (Monthly)', fontsize=13)
ax.set_ylabel('Deals Closed (Monthly)', fontsize=13)
ax.legend(frameon=False, fontsize=11)
ax.grid(True, alpha=0.3, color='#DDDDDD')
ax.spines[['top', 'right']].set_visible(False)

fig.text(0.1, -0.03,
    '✓ CORRECTED: 79 reps in quiet neutral gray (70%), trend line in muted lavender (20%), '
    'one orange accent on the actionable outlier (10%).',
    fontsize=10, color=PALETTE['green'], style='italic')

fig.text(0.1, -0.06, 'Source: Simulated sales rep performance data',
         fontsize=9, color='#888', style='italic')

plt.tight_layout()
plt.show()
````

### Demo — Business Dashboard Mockup Implementing 70-20-10

The following chart shows a realistic executive KPI summary. The entire layout is built around the 70-20-10 framework: a quiet neutral base, one brand-color supporting metric, and one gold accent for the record-breaking data point.

````python
# ── DEMO: Business dashboard mockup with explicit 70-20-10 labeling ─────────────
quarters  = ['Q1', 'Q2', 'Q3', 'Q4']
q_revenue = [14.4, 17.1, 21.3, 24.6]   # Q4 is the record
q_target  = [15.0, 16.0, 20.0, 22.0]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor(PALETTE['bg'])  # 70% base background
fig.suptitle('FY Annual Revenue Performance Dashboard',
             fontsize=17, fontweight='bold', color=PALETTE['primary'])

# ── Left panel: quarterly revenue bars ──────────────────────────────────────────
ax = axes[0]
ax.set_facecolor('white')  # 70%

bar_colors = [PALETTE['neutral'] if q != 'Q4' else PALETTE['gold'] for q in quarters]
bars = ax.bar(quarters, q_revenue, color=bar_colors, width=0.5,
              edgecolor='white', linewidth=0.8, zorder=3)
label_colors = [PALETTE['primary'] if c == PALETTE['gold'] else '#888' for c in bar_colors]
labels = ax.bar_label(bars, fmt='$%.1fM', fontsize=11, padding=4, fontweight='bold')
for lbl, lc in zip(labels, label_colors):
    lbl.set_color(lc)

# Target line — 20% supporting element
ax.step([-.4, .6, 1.6, 2.6, 3.6], [q_target[0]] + q_target,
        where='post', color=PALETTE['lavender'], linewidth=2,
        linestyle='--', label='Quarterly target', zorder=4)

ax.set_title('Quarterly Revenue vs. Target', fontsize=13, fontweight='bold', color=PALETTE['primary'])
ax.set_ylabel('Revenue ($M)', fontsize=12)
ax.set_ylim(0, 30)
ax.legend(fontsize=10, frameon=False)
ax.grid(True, axis='y', alpha=0.35, color='#DDDDDD', zorder=0)
ax.spines[['top', 'right']].set_visible(False)
ax.annotate('Record quarter\n+12% vs target',
            xy=(3, q_revenue[3]), xytext=(2.3, q_revenue[3] + 2),
            fontsize=9, color=PALETTE['gold'], fontweight='bold',
            arrowprops=dict(arrowstyle='->', color=PALETTE['gold'], lw=1.2))

# ── Right panel: monthly revenue line with accent on record month ────────────────
ax2 = axes[1]
ax2.set_facecolor('white')  # 70%

# 70%: plot all points as neutral gray
ax2.plot(range(12), revenue, color=PALETTE['neutral'], linewidth=2, zorder=2)
ax2.scatter(range(12), revenue,
            c=[PALETTE['gold'] if r == max(revenue) else PALETTE['neutral'] for r in revenue],
            s=[140 if r == max(revenue) else 60 for r in revenue],
            edgecolors='white', linewidth=0.8, zorder=4)

# 20%: expense comparison in muted lavender
ax2.plot(range(12), expenses, color=PALETTE['lavender'], linewidth=1.8,
         linestyle='--', label='Expenses', zorder=2)

# 10%: target line as accent dashed red
ax2.axhline(y=7.0, color=PALETTE['red'], linewidth=1.5,
            linestyle=':', alpha=0.8, label='Monthly target $7M', zorder=3)

ax2.set_title('Monthly Revenue & Expenses ($M)', fontsize=13,
              fontweight='bold', color=PALETTE['primary'])
ax2.set_ylabel('Amount ($M)', fontsize=12)
ax2.set_xticks(range(12))
ax2.set_xticklabels(months, rotation=30, fontsize=9)
ax2.legend(fontsize=10, frameon=False)
ax2.grid(True, alpha=0.35, color='#DDDDDD', zorder=0)
ax2.spines[['top', 'right']].set_visible(False)

peak_m = revenue.index(max(revenue))
ax2.annotate(f'Dec — ${max(revenue)}M',
             xy=(peak_m, max(revenue)),
             xytext=(peak_m - 3.5, max(revenue) - 0.8),
             fontsize=9, color=PALETTE['gold'], fontweight='bold',
             arrowprops=dict(arrowstyle='->', color=PALETTE['gold'], lw=1.2))

# Color role legend
legend_elements = [
    mpatches.Patch(facecolor=PALETTE['neutral'],  label='70% Neutral base'),
    mpatches.Patch(facecolor=PALETTE['lavender'], label='20% Supporting'),
    mpatches.Patch(facecolor=PALETTE['gold'],     label='10% Accent (KPI)'),
]
fig.legend(handles=legend_elements, loc='lower center', ncol=3,
           fontsize=10, frameon=True, fancybox=True,
           bbox_to_anchor=(0.5, -0.08))

fig.text(0.1, -0.14, 'Source: Simulated annual KPI data',
         fontsize=9, color='#888', style='italic')

plt.tight_layout()
plt.show()
````

### Demo — Palette Consistency Across Three Related Charts

The following three charts cover different aspects of the same sales dataset — by region, by product line, and over time. Despite showing different chart types, they feel like a single coherent report because every color assignment is identical across all three panels.

````python
# ── DEMO: Consistent palette across 3 related charts ────────────────────────────

# Shared category → color mapping (NEVER reassign mid-report)
CAT_COLORS = {
    'North':  PALETTE['primary'],
    'South':  PALETTE['lavender'],
    'East':   PALETTE['neutral'],
    'West':   PALETTE['neutral2'],
}

# Simulated data
np.random.seed(3)
regions    = list(CAT_COLORS.keys())
annual_rev = {'North': 38.4, 'South': 24.1, 'East': 17.9, 'West': 12.6}

monthly_by_region = {
    r: (np.random.normal(annual_rev[r] / 12, annual_rev[r] / 60, 12)).clip(0)
    for r in regions
}

q_by_region = {
    r: [monthly_by_region[r][i*3:(i+1)*3].sum() for i in range(4)]
    for r in regions
}

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.patch.set_facecolor(PALETTE['bg'])
fig.suptitle('Regional Sales Performance — FY Report',
             fontsize=17, fontweight='bold', color=PALETTE['primary'])

# Chart 1: Annual revenue bar chart
ax1 = axes[0]
ax1.set_facecolor('white')
ax1.bar(regions, [annual_rev[r] for r in regions],
        color=[CAT_COLORS[r] for r in regions],
        edgecolor='white', linewidth=0.8)
ax1.set_title('Annual Revenue by Region', fontsize=13,
              fontweight='bold', color=PALETTE['primary'])
ax1.set_ylabel('Revenue ($M)', fontsize=12)
ax1.grid(True, axis='y', alpha=0.3, color='#DDDDDD')
ax1.spines[['top', 'right']].set_visible(False)

# Chart 2: Quarterly breakdown — stacked bar
ax2 = axes[1]
ax2.set_facecolor('white')
bottom = np.zeros(4)
for r in regions:
    ax2.bar(quarters, q_by_region[r], bottom=bottom,
            color=CAT_COLORS[r], label=r, edgecolor='white', linewidth=0.5)
    bottom += np.array(q_by_region[r])
ax2.set_title('Quarterly Revenue (Stacked)', fontsize=13,
              fontweight='bold', color=PALETTE['primary'])
ax2.set_ylabel('Revenue ($M)', fontsize=12)
ax2.legend(fontsize=10, frameon=False, loc='upper left')
ax2.grid(True, axis='y', alpha=0.3, color='#DDDDDD')
ax2.spines[['top', 'right']].set_visible(False)

# Chart 3: Monthly trend lines
ax3 = axes[2]
ax3.set_facecolor('white')
for r in regions:
    lw = 2.5 if r == 'North' else 1.5
    ax3.plot(range(12), monthly_by_region[r], color=CAT_COLORS[r],
             linewidth=lw, marker='o', markersize=4, label=r)
ax3.set_title('Monthly Revenue Trend', fontsize=13,
              fontweight='bold', color=PALETTE['primary'])
ax3.set_ylabel('Revenue ($M)', fontsize=12)
ax3.set_xticks(range(12))
ax3.set_xticklabels(months, rotation=45, fontsize=8)
ax3.legend(fontsize=10, frameon=False)
ax3.grid(True, alpha=0.3, color='#DDDDDD')
ax3.spines[['top', 'right']].set_visible(False)

fig.text(0.5, -0.04,
    '✓ CONSISTENT: North is always #4B2E83, South is always #B7A6D5, '
    'East and West always in neutral gray — across all three chart types.',
    ha='center', fontsize=10, color=PALETTE['green'], style='italic')

fig.text(0.1, -0.08, 'Source: Simulated regional sales data',
         fontsize=9, color='#888', style='italic')

plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">1. Truncated Y-Axis</h2></div>

**What it is:** The y-axis starts well above zero, so differences that are small in absolute terms appear enormous.

**Why it deceives:** Human perception encodes magnitude through bar height. A bar twice as tall reads as "twice as large" — even when the actual difference is 3 %.

**The fix:** Start bar-chart y-axes at 0. When a zoomed view genuinely helps (e.g., showing subtle variation in a line chart), label the axis break clearly with a jagged-line symbol and note the truncation in the caption.

> **Classic example:** Fox News repeatedly aired bar charts of tax rates with y-axes starting at 33 %, making a rate change from 35 % to 39.6 % look like a tripling.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">2. Manipulated Scales (Non-Linear Without Disclosure)</h2></div>

**What it is:** A logarithmic or otherwise non-linear axis is used without any label or warning, compressing large values and making exponential growth look linear.

**Why it deceives:** Readers assume linear spacing by default. On a log scale, the visual distance from 10 to 100 equals the distance from 100 to 1,000 — a 10× difference looks identical to a 900× difference.

**The fix:** Label log axes explicitly (e.g., "log scale" in the axis title). Add a note in the caption. When in doubt, provide both a linear and a log panel side by side.

> **When log scales ARE appropriate:** Displaying data spanning multiple orders of magnitude (e.g., COVID case counts during an exponential phase), or comparing growth rates rather than absolute levels.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">3. Cherry-Picked Date Windows</h2></div>

**What it is:** The chart shows only a subset of the available time range — the subset that best supports the desired narrative.

**Why it deceives:** A three-month rally in a flat or declining annual trend looks like proof of success. The reader has no reference point for what happened before or after the window.

**The fix:** Always show the full available history, or explicitly justify and label any zoom window. A small inset panel showing the full range with the zoomed region highlighted is best practice.

> **Rule:** If you chose your date range *after* looking at the data, you are almost certainly cherry-picking. Pre-register your time window before visualising.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">4. Misleading Cumulative Charts</h2></div>

**What it is:** Plotting cumulative totals (always increasing by definition) instead of period-by-period values to mask a slowdown or decline.

**Why it deceives:** A cumulative line never goes down. New signups, revenue, or cases can be dropping week-over-week, but the cumulative curve still rises — creating a false impression of continued growth.

**The fix:** Show both the cumulative total *and* the periodic (weekly/monthly) change. When reporting growth, use the periodic chart as the primary view and reserve cumulative charts for contexts where total size is the meaningful metric (e.g., total vaccines administered).

> **Tip:** Always ask: *Is the question about total stock or about flow?* Cumulative = stock. Period-over-period = flow.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">5. Dual-Axis Deception</h2></div>

**What it is:** Two completely unrelated variables are plotted on a single chart using different left and right y-axes. By rescaling either axis, the maker can manufacture visual correlation between any two time series.

**Why it deceives:** Lines that cross or run in parallel trigger a strong correlation inference in the reader's brain. The axes are freely rescalable, so *any* two trending series can be made to look related.

**The fix:** Use dual axes only when both variables share a meaningful relationship (e.g., temperature in °C and °F, or price and volume of the *same* asset). For genuinely separate metrics, use two side-by-side charts with consistent x-axes.

> **Famous example:** Charts showing that Nicolas Cage films and US pool drownings both rose together in the 2000s — a classic spurious correlation.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">6. Distorted Aspect Ratios</h2></div>

**What it is:** Making a chart very wide and short (or very tall and narrow) to flatten or exaggerate the slope of a trend line.

**Why it deceives:** A 45° slope is the most visually neutral representation of a trend. Extreme aspect ratios turn gradual changes into alarmist spikes, or dramatic changes into boring flatlines.

**The fix:** Follow Cleveland's (1993) *banking to 45°* principle: adjust the aspect ratio so that the average absolute slope of the line segments is approximately 45°. Matplotlib's `ax.set_aspect('equal')` or manual figure sizing helps.

> **Practical rule:** If your trend line hugs either the top or bottom of the chart, your aspect ratio is likely misleading.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">7. 3D Distortion</h2></div>

**What it is:** Adding a false third dimension (perspective/depth) to a 2D chart — most commonly pie charts and bar charts — purely for aesthetic effect.

**Why it deceives:** Perspective projection changes the apparent size of elements based on their position. Slices at the front of a 3D pie chart look larger than equally-sized slices at the back. Bars near the viewer appear taller.

**The fix:** Never use 3D effects on charts that encode data in 2D. Replace 3D pie charts with flat pie charts or, better, horizontal bar charts. Remove all decorative depth.

> **Edward Tufte** called chartjunk — including 3D effects — "the interior decoration of statistics": it increases ink without increasing information.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">8. Area Exaggeration (Scaling by Radius Instead of Area)</h2></div>

**What it is:** When using circles or icons to represent quantities, the *radius* (or side length) is scaled proportionally to the value rather than the *area* — causing the larger symbol to be dramatically over-sized.

**Why it deceives:** If Country A has twice the GDP of Country B and you double the radius of its circle, its *area* is four times larger (area = π r²), making the difference look four-fold rather than two-fold.

**The fix:** Scale the area, not the radius. In matplotlib, pass `s = value` to `ax.scatter()` (bubble area) or compute `radius = sqrt(value / pi)` manually. Always include a size legend with labelled reference bubbles.

> **Rule:** `marker_size ∝ sqrt(value)` for radius-based markers; `marker_area ∝ value`.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">9. Overloaded Pie Charts</h2></div>

**What it is:** A pie chart with too many slices (often 7 or more), many of which are similarly sized, forcing readers to rely on labels rather than visual angles.

**Why it deceives:** Human perception of angles is poor; we cannot reliably distinguish between a 12 % slice and a 15 % slice. When many thin slices cluster together, the chart communicates almost nothing visually.

**The fix:** Use a pie chart only for 2–5 categories where one or two slices clearly dominate. For more categories, use a horizontal bar chart sorted by magnitude — bar length is a far more accurate perceptual encoding than angle.

> **Guideline:** If you need a legend to read a pie chart, it's probably the wrong chart type.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">10. Bubble Chart Misuse</h2></div>

**What it is:** Bubble charts where the size encoding is inconsistent, unlabelled, or scaled by radius rather than area — so the visual comparison between bubbles is meaningless.

**Why it deceives:** Without a size legend that shows what a given bubble area means, readers cannot translate visual size into a data value. Combined with radius scaling (see #8), bubbles become pure decoration.

**The fix:** Always include a size legend with at least three reference bubbles at known values. Scale by area. Choose bubble sizes so the chart is not visually dominated by a single huge bubble that obscures the others.

> **Extra caution:** Overlapping bubbles should use transparency (`alpha`) and be clearly acknowledged in the caption.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">11. Misleading Choropleth Maps (Population vs. Per-Capita)</h2></div>

**What it is:** A choropleth map colours regions by a raw count (e.g., total COVID cases, total crime incidents) rather than a rate (cases per 100,000 people), making densely populated regions look like outliers simply because they have more people.

**Why it deceives:** Large, sparsely populated states dominate the eye by area even if they have low per-capita rates. Dense urban states look "worse" on raw-count maps for any metric that scales with population.

**The fix:** Normalise by population (or land area, or some other denominator appropriate to the question). Always state in the title whether you are showing a count or a rate. When both views are relevant, show them side by side.

> **Classic trap:** Maps of total opioid deaths by US state make rural states look safe. Per-capita maps tell a very different story.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">12. Unequal Bin Widths in Histograms</h2></div>

**What it is:** A histogram where bins are not all the same width, but the y-axis shows *frequency* (count) rather than *frequency density* (count ÷ bin width). Wide bins look like they contain more observations than they actually do per unit of the x-axis.

**Why it deceives:** A bin of width 20 that spans ages 60–80 and contains 40 people looks as "tall" as a bin of width 5 spanning ages 20–25 that contains 40 people — but the older-age group is actually 4× less dense per year.

**The fix:** If you use unequal bins, plot *density* on the y-axis (use `density=True` in matplotlib). Better still, use equal-width bins unless there is a strong domain reason for variable widths (e.g., age-standardised epidemiological tables). Always label what the y-axis represents.

> **Matplotlib tip:** `ax.hist(data, bins=[...], density=True)` automatically converts to frequency density.

The four bars below show quarterly revenue for a retail company. The y-axis starts at **95** instead of **0**. The difference between Q1 and Q4 is just **3 %** — but the chart makes Q4 look roughly **10x** taller than Q1, implying an explosive growth story that does not exist in the data.

````python
# Simulated quarterly revenue (in $M) — a 3 % increase from Q1 to Q4
quarters = ['Q1 2024', 'Q2 2024', 'Q3 2024', 'Q4 2024']
revenue  = [97.2, 98.1, 99.0, 100.2]   # range: 97.2 to 100.2 — a ~3 % spread

fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.bar(quarters, revenue, color='#C62828', edgecolor='white', linewidth=1.5)

# ---- THE DECEPTION: y-axis starts at 95, NOT 0 ----
ax.set_ylim(95, 102)

for bar, val in zip(bars, revenue):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
            f'${val:.1f}M', ha='center', va='bottom', fontsize=11, fontweight='bold')

ax.set_title('Quarterly Revenue — MISLEADING (truncated axis)',
             fontsize=14, fontweight='bold', color='#C62828')
ax.set_xlabel('Quarter', fontsize=12)
ax.set_ylabel('Revenue ($M)  \u26a0 y-axis starts at 95!', fontsize=12, color='#C62828')
ax.grid(True, alpha=0.3)

fig.text(0.1, -0.02, 'Source: Simulated retail revenue data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()

print(f'Actual Q1-to-Q4 change: +${revenue[-1]-revenue[0]:.1f}M  '
      f'= +{(revenue[-1]/revenue[0]-1)*100:.1f}%')
print('Yet the bar for Q4 appears ~10x taller than Q1 in this chart!')
````

````python
fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.bar(quarters, revenue, color='#388E3C', edgecolor='white', linewidth=1.5)

# ---- THE FIX: y-axis starts at 0 ----
ax.set_ylim(0, 115)

for bar, val in zip(bars, revenue):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.8,
            f'${val:.1f}M', ha='center', va='bottom', fontsize=11, fontweight='bold')

# Annotate the actual change
pct_change = (revenue[-1] / revenue[0] - 1) * 100
ax.annotate(
    f'Q1 \u2192 Q4: +{pct_change:.1f}% actual growth',
    xy=(3, revenue[-1]), xytext=(2.2, 108),
    arrowprops=dict(arrowstyle='->', color='#333', lw=1.5),
    fontsize=11, color='#333',
    bbox=dict(boxstyle='round,pad=0.3', facecolor='#E8F5E9', edgecolor='#388E3C')
)

ax.set_title('Quarterly Revenue — CORRECTED (y-axis from 0)',
             fontsize=14, fontweight='bold', color='#388E3C')
ax.set_xlabel('Quarter', fontsize=12)
ax.set_ylabel('Revenue ($M)', fontsize=12)
ax.grid(True, alpha=0.3)

fig.text(0.1, -0.02, 'Source: Simulated retail revenue data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()
````

This line chart shows monthly online sales for an e-commerce company. Only the **3 best consecutive months** (April–June) are displayed, making the business look like it has strong upward momentum — when the full annual picture tells a very different story.

<div style="border-left: 5px solid #C62828; background-color: #FFF8F8; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #C62828; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Demo 3 — Dual-Axis Deception vs. Honest Side-by-Side</h2></div>

Two completely unrelated metrics — a company's quarterly **ad spending** and the **average outdoor temperature** in their city — are forced onto a dual-axis chart. By rescaling the right axis, we can make them appear perfectly correlated. The corrected version uses two separate panels.

````python
quarters_dual = ['Q1', 'Q2', 'Q3', 'Q4']
ad_spend      = [120, 185, 210, 140]   # $K — peaks in summer campaign season
avg_temp      = [38,   62,  78,  45]   # °F  — seasonal temperature, entirely unrelated

# ---- MISLEADING: dual-axis implies correlation ----
fig, ax1 = plt.subplots(figsize=(9, 6))
ax2 = ax1.twinx()

l1, = ax1.plot(quarters_dual, ad_spend, marker='o', linewidth=2.5, markersize=9,
               color='#4B2E83', label='Ad Spend ($K)')
l2, = ax2.plot(quarters_dual, avg_temp, marker='s', linewidth=2.5, markersize=9,
               color='#E8B923', linestyle='--', label='Avg Temp (°F)')

# Scale right axis to align the lines visually
ax2.set_ylim(20, 100)
ax1.set_ylim(80, 260)

ax1.set_xlabel('Quarter', fontsize=12)
ax1.set_ylabel('Ad Spend ($K)', fontsize=12, color='#4B2E83')
ax2.set_ylabel('Avg Temperature (°F)', fontsize=12, color='#E8B923')
ax1.set_title('Ad Spend & Temperature — MISLEADING (dual axis implies correlation)',
              fontsize=12, fontweight='bold', color='#C62828')

lines = [l1, l2]
ax1.legend(lines, [l.get_label() for l in lines],
           frameon=True, fancybox=True, shadow=True, fontsize=10)
ax1.grid(True, alpha=0.3)

fig.text(0.1, -0.02, 'Source: Simulated marketing and weather data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()
print('These two variables have NO causal relationship. The dual axis manufactured the visual correlation.')
````

````python
# ---- CORRECTED: two honest side-by-side panels ----
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(13, 5))

ax_left.plot(quarters_dual, ad_spend, marker='o', linewidth=2.5,
             markersize=9, color='#4B2E83')
ax_left.set_title('Ad Spend per Quarter', fontsize=13, fontweight='bold', color='#4B2E83')
ax_left.set_xlabel('Quarter', fontsize=12)
ax_left.set_ylabel('Ad Spend ($K)', fontsize=12)
ax_left.set_ylim(0, 250)
ax_left.grid(True, alpha=0.3)

ax_right.plot(quarters_dual, avg_temp, marker='s', linewidth=2.5,
              markersize=9, color='#E8B923', linestyle='--')
ax_right.set_title('Average Temperature per Quarter', fontsize=13, fontweight='bold', color='#85714D')
ax_right.set_xlabel('Quarter', fontsize=12)
ax_right.set_ylabel('Avg Temperature (\u00b0F)', fontsize=12)
ax_right.set_ylim(0, 100)
ax_right.grid(True, alpha=0.3)

fig.suptitle('CORRECTED: Separate panels — no implied correlation',
             fontsize=14, fontweight='bold', color='#388E3C', y=1.03)
fig.text(0.1, -0.03, 'Source: Simulated marketing and weather data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #C62828; background-color: #FFF8F8; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #C62828; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Demo 4 — 3D Pie Chart vs. Clean Horizontal Bar Chart</h2></div>

A 3D perspective pie chart distorts the apparent size of slices. The front slice ("West") looks larger than it is; the back slices look smaller. The horizontal bar chart below communicates the same information accurately.

````python
regions  = ['Northeast', 'Southeast', 'Midwest', 'West', 'Southwest']
shares   = [22, 18, 20, 25, 15]   # market share %
explode  = (0, 0, 0, 0.12, 0)     # pull West slice to the front

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ---- LEFT: 3D-style (wedge shadows + explode + tilt illusion) ----
colors_pie = ['#4B2E83', '#E8B923', '#85714D', '#C62828', '#1976D2']

wedges, texts, autotexts = axes[0].pie(
    shares, labels=regions, colors=colors_pie, autopct='%1.0f%%',
    explode=explode, startangle=200, pctdistance=0.6,
    shadow=True,   # <-- the 3D illusion
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
for t in autotexts:
    t.set_fontweight('bold'); t.set_color('white'); t.set_fontsize(11)
axes[0].set_title('Market Share by Region\nMISLEADING: 3D distortion exaggerates \'West\'',
                  fontsize=11, fontweight='bold', color='#C62828')

# ---- RIGHT: Honest horizontal bar chart ----
sorted_idx = np.argsort(shares)
sorted_regions = [regions[i] for i in sorted_idx]
sorted_shares  = [shares[i]  for i in sorted_idx]
bar_colors     = [colors_pie[i] for i in sorted_idx]

axes[1].barh(sorted_regions, sorted_shares, color=bar_colors, edgecolor='white', linewidth=1.5)
for i, val in enumerate(sorted_shares):
    axes[1].text(val + 0.3, i, f'{val}%', va='center', fontsize=11, fontweight='bold')
axes[1].set_xlim(0, 32)
axes[1].set_xlabel('Market Share (%)', fontsize=12)
axes[1].set_title('Market Share by Region\nCORRECTED: Horizontal bar (accurate lengths)',
                  fontsize=11, fontweight='bold', color='#388E3C')
axes[1].grid(True, axis='x', alpha=0.3)
axes[1].axvline(20, color='#333', linewidth=0.8, linestyle=':', alpha=0.5)

fig.text(0.1, -0.02, 'Source: Simulated retail market share data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()
````

<div style="border-left: 5px solid #C62828; background-color: #FFF8F8; padding: 20px 25px; margin: 20px 0; border-radius: 5px;"><h2 style="color: #C62828; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Demo 5 — Unequal Histogram Bins That Hide the True Distribution</h2></div>

The left histogram uses deliberately unequal bin widths (wide bins in the upper tail) while plotting raw counts — making the distribution look bimodal. The right histogram uses equal-width bins and reveals the true right-skewed shape.

````python
# Simulate patient BMI data — log-normal, right-skewed
np.random.seed(7)
bmi = np.random.lognormal(mean=3.3, sigma=0.18, size=800)  # BMI roughly 18–45
bmi = bmi[(bmi >= 15) & (bmi <= 55)]

# Unequal bins: fine resolution in the centre, coarse in the tails
unequal_bins = [15, 18, 20, 22, 24, 26, 28, 30, 35, 40, 55]

fig, (ax_bad, ax_good) = plt.subplots(1, 2, figsize=(14, 6))

# ---- LEFT: unequal bins with COUNT on y-axis ----
ax_bad.hist(bmi, bins=unequal_bins, color='#C62828', edgecolor='white', linewidth=0.8)
ax_bad.set_title('BMI Distribution — MISLEADING\n(unequal bins, count on y-axis)',
                 fontsize=12, fontweight='bold', color='#C62828')
ax_bad.set_xlabel('BMI', fontsize=12)
ax_bad.set_ylabel('Count  (not density!)', fontsize=12, color='#C62828')

# Mark the wide tail bin
ax_bad.axvspan(40, 55, alpha=0.15, color='#F57C00',
               label='Width = 15  \u26a0 inflated count')
ax_bad.legend(fontsize=9)
ax_bad.grid(True, alpha=0.3)

bin_edges = []
for start, end in zip(unequal_bins[:-1], unequal_bins[1:]):
    ax_bad.annotate(f'{end-start}', xy=((start+end)/2, 5),
                    ha='center', fontsize=8, color='#555')

# ---- RIGHT: equal bins with DENSITY on y-axis ----
ax_good.hist(bmi, bins=30, density=True, color='#388E3C', edgecolor='white', linewidth=0.8, alpha=0.85)
ax_good.set_title('BMI Distribution — CORRECTED\n(equal bins, density on y-axis)',
                  fontsize=12, fontweight='bold', color='#388E3C')
ax_good.set_xlabel('BMI', fontsize=12)
ax_good.set_ylabel('Density', fontsize=12)
ax_good.axvline(np.median(bmi), color='#C62828', linewidth=2, linestyle='--',
                label=f'Median BMI = {np.median(bmi):.1f}')
ax_good.legend(fontsize=10)
ax_good.grid(True, alpha=0.3)

fig.text(0.1, -0.02, 'Source: Simulated patient BMI data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()
````

````python
# Exercise 12 — Starter code (DO NOT MODIFY THIS CELL)
np.random.seed(99)

# Simulate recovery times (days) — log-normal: most patients recover in 3-10 days,
# a minority take much longer (right-skewed)
recovery_days = np.random.lognormal(mean=1.9, sigma=0.5, size=600)
recovery_days = recovery_days[(recovery_days >= 1) & (recovery_days <= 45)]

# Unequal bins: fine resolution 1-15, one huge bin 15-45
unequal_bins_ex12 = [1, 3, 5, 7, 9, 11, 13, 15, 45]

fig, ax = plt.subplots(figsize=(10, 6))

n_vals, bin_edges, patches = ax.hist(
    recovery_days, bins=unequal_bins_ex12,
    color='#7B1FA2', edgecolor='white', linewidth=0.8
    # Note: density=False — counts on y-axis
)

# Add fake 3D shadow effect
for patch in patches:
    shadow = mpatches.FancyBboxPatch(
        (patch.get_x() + 0.15, patch.get_y() - 3),
        patch.get_width(), patch.get_height(),
        boxstyle='square', linewidth=0,
        facecolor='#BDB0C9', zorder=0, alpha=0.6
    )
    ax.add_patch(shadow)

# Annotate bin widths
for left, right, count in zip(bin_edges[:-1], bin_edges[1:], n_vals):
    mid = (left + right) / 2
    ax.text(mid, count + 2, f'w={int(right-left)}', ha='center',
            fontsize=8, color='#444')

ax.set_title('Patient Recovery Time — MISLEADING\n(unequal bins + counts + 3D shadow)',
             fontsize=13, fontweight='bold', color='#C62828')
ax.set_xlabel('Recovery Time (days)', fontsize=12)
ax.set_ylabel('Count  (not density!)', fontsize=12, color='#C62828')
ax.grid(True, alpha=0.3)

fig.text(0.1, -0.02, 'Source: Simulated post-surgical recovery data',
         fontsize=9, color='#888', style='italic')
plt.tight_layout()
plt.show()

print(f'Wide tail bin (15-45 days): width = 30, count = {int(n_vals[-1])}')
print(f'Narrow bins (e.g. 1-3 days): width = 2, count = {int(n_vals[0])}')
print(f'Density in tail bin: {n_vals[-1]/30:.1f} per day vs narrow bin: {n_vals[0]/2:.1f} per day')
````

<div style="background-color: #E8F5E9; border-left: 4px solid #388E3C; padding: 15px 20px; margin: 20px 0; border-radius: 5px;"><h3 style="color: #388E3C; margin-top: 0;">Demo 1 (Corrected) — Absolute Numbers, Sample Size, and Context</h3><p style="color: #333; margin-bottom: 0;">The corrected version shows both absolute counts and the sample size denominator, labels the baseline explicitly, and adds a subtitle that prevents headline inflation.</p></div>

````python
n_total  = 10
before   = 2
after    = 8
rel_chg  = (after - before) / before * 100
abs_chg  = after - before

fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Left panel — absolute counts with denominator
ax = axes[0]
bars = ax.bar(['Before', 'After'], [before, after],
              color=[LAVENDER, PRIMARY], width=0.5,
              edgecolor='white', linewidth=1.5)
ax.set_title(f'Feature Adoption (n = {n_total} users)', fontsize=13,
             fontweight='bold', color=PRIMARY)
ax.set_ylabel('Users who adopted feature', fontsize=11)
ax.set_ylim(0, n_total + 1)
# Add a dashed line at total n for reference
ax.axhline(n_total, color=TAN, linestyle='--', linewidth=1.2, label=f'Total users = {n_total}')
ax.legend(fontsize=10)
for bar, val in zip(bars, [before, after]):
    pct = val / n_total * 100
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.15,
            f'{val} ({pct:.0f}%)', ha='center', va='bottom',
            fontweight='bold', fontsize=12)
ax.set_xlabel(
    f'Relative change: +{rel_chg:.0f}%  |  Absolute change: +{abs_chg} users',
    fontsize=10, color='#555')

# Right panel — adoption rate with 95% Wilson CI
ax2 = axes[1]
from statsmodels.stats.proportion import proportion_confint
ci_before = proportion_confint(before, n_total, method='wilson')
ci_after  = proportion_confint(after,  n_total, method='wilson')
rates  = [before / n_total, after  / n_total]
errs_lo = [rates[0] - ci_before[0], rates[1] - ci_after[0]]
errs_hi = [ci_before[1] - rates[0], ci_after[1]  - rates[1]]
ax2.bar(['Before', 'After'], rates,
        color=[LAVENDER, PRIMARY], width=0.5,
        edgecolor='white', linewidth=1.5, yerr=[errs_lo, errs_hi],
        capsize=8, error_kw=dict(ecolor='#555', lw=2))
ax2.set_title('Adoption Rate with 95% Confidence Interval', fontsize=13,
              fontweight='bold', color=PRIMARY)
ax2.set_ylabel('Proportion of users', fontsize=11)
ax2.set_ylim(0, 1.15)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
ax2.text(0.5, 1.05,
         'Wide CIs reveal the estimate is highly uncertain at n = 10',
         ha='center', transform=ax2.transAxes, fontsize=10,
         color='#555', style='italic')

plt.suptitle('Honest reporting: same data, full context', fontsize=14,
             fontweight='bold', color=GREEN, y=1.01)
plt.tight_layout()
plt.show()
print('Key takeaway: a 300% relative increase from 2 to 8 users (out of 10)')
print('is statistically meaningless. The 95% CI for the "after" rate spans')
print('roughly 45%–98%, making any strong claim premature.')
````

<div style="background-color: #E3F2FD; border-left: 4px solid #1976D2; padding: 15px 20px; margin: 20px 0; border-radius: 5px;"><h3 style="color: #1976D2; margin-top: 0;">Demo 3 — Scatter Plot Implying Causality: Ice Cream Sales vs Drowning Deaths</h3><p style="color: #333; margin-bottom: 0;">Both variables rise in summer. The correlation is real — but <strong>neither causes the other</strong>. The confounding variable is <em>ambient temperature / season</em>, which independently drives both ice cream consumption and swimming activity.</p></div>

````python
np.random.seed(21)

# Confound: temperature index (0 = winter, 1 = summer)
months  = np.arange(1, 13)
temp    = np.sin((months - 3) * np.pi / 6) * 0.5 + 0.5   # 0..1 seasonal wave

ice_cream = 100 + 300 * temp + np.random.normal(0, 20, 12)
drownings =  10 +  50 * temp + np.random.normal(0,  4, 12)

r, p = stats.pearsonr(ice_cream, drownings)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left — deceptive scatter
ax = axes[0]
ax.scatter(ice_cream, drownings, color=PRIMARY, s=80, zorder=3)
m_slope, m_int, *_ = stats.linregress(ice_cream, drownings)
x_line = np.linspace(ice_cream.min(), ice_cream.max(), 200)
ax.plot(x_line, m_slope * x_line + m_int, color=RED, linewidth=2,
        label=f'r = {r:.2f}, p = {p:.3f}')
ax.set_xlabel('Ice Cream Sales (units / month)', fontsize=11)
ax.set_ylabel('Drowning Deaths per Month', fontsize=11)
ax.set_title('"Ice Cream Causes Drowning?"\nDeceptive framing', fontsize=13,
             fontweight='bold', color=RED)
ax.legend(fontsize=10)
ax.text(0.05, 0.93, 'Strong positive correlation!',
        transform=ax.transAxes, color=RED, fontsize=10, style='italic')

# Right — honest version showing confounder
ax2 = axes[1]
sc = ax2.scatter(ice_cream, drownings, c=temp, cmap='RdPu', s=100,
                 zorder=3, edgecolors='#444', linewidths=0.5)
cbar = plt.colorbar(sc, ax=ax2)
cbar.set_label('Temperature Index (0=cold, 1=hot)', fontsize=10)
for i, m in enumerate(months):
    import calendar
    ax2.annotate(calendar.month_abbr[m],
                 (ice_cream[i], drownings[i]),
                 textcoords='offset points', xytext=(5, 4), fontsize=8)
ax2.set_xlabel('Ice Cream Sales (units / month)', fontsize=11)
ax2.set_ylabel('Drowning Deaths per Month', fontsize=11)
ax2.set_title('Honest framing: colour reveals the\nconfounding variable (temperature)',
              fontsize=13, fontweight='bold', color=GREEN)
ax2.text(0.05, 0.93,
         'Both variables are driven by summer heat,\nnot by each other.',
         transform=ax2.transAxes, color='#333', fontsize=9, style='italic')

plt.suptitle('Correlation does NOT imply causation', fontsize=14,
             fontweight='bold', color=PRIMARY, y=1.01)
plt.tight_layout()
plt.show()
print(f'Pearson r = {r:.3f}  (p = {p:.4f}) — strong correlation, zero causal link.')
````

<div style="background-color: #E3F2FD; border-left: 4px solid #1976D2; padding: 15px 20px; margin: 20px 0; border-radius: 5px;"><h3 style="color: #1976D2; margin-top: 0;">Demo 4 — Bar Chart Hiding Overlapping Confidence Intervals</h3><p style="color: #333; margin-bottom: 0;">Two bars that <em>look</em> different in height can have completely overlapping 95% confidence intervals, meaning the difference is not statistically significant. Omitting error bars is one of the most common ways to manufacture apparent differences.</p></div>

````python
np.random.seed(3)

n_a, n_b = 30, 30
group_a  = np.random.normal(52, 18, n_a)
group_b  = np.random.normal(58, 20, n_b)

mean_a, se_a = group_a.mean(), stats.sem(group_a)
mean_b, se_b = group_b.mean(), stats.sem(group_b)
ci_a = stats.t.interval(0.95, df=n_a-1, loc=mean_a, scale=se_a)
ci_b = stats.t.interval(0.95, df=n_b-1, loc=mean_b, scale=se_b)

fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Bad — no error bars
ax = axes[0]
ax.bar(['Group A', 'Group B'], [mean_a, mean_b],
       color=[LAVENDER, PRIMARY], width=0.5, edgecolor='white')
ax.set_title('"Group B scores higher!"\n(No uncertainty shown)', fontsize=13,
             fontweight='bold', color=RED)
ax.set_ylabel('Score', fontsize=12)
ax.set_ylim(0, 80)
for x, val in enumerate([mean_a, mean_b]):
    ax.text(x, val + 0.8, f'{val:.1f}', ha='center', fontweight='bold', fontsize=12)

# Good — with 95% CI
ax2 = axes[1]
errs  = [[mean_a - ci_a[0], mean_b - ci_b[0]],
          [ci_a[1] - mean_a, ci_b[1] - mean_b]]
ax2.bar(['Group A', 'Group B'], [mean_a, mean_b],
        color=[LAVENDER, PRIMARY], width=0.5, edgecolor='white',
        yerr=errs, capsize=9,
        error_kw=dict(ecolor='#333', lw=2, capthick=2))
ax2.set_title('With 95% Confidence Intervals\n(Overlap is visible)', fontsize=13,
              fontweight='bold', color=GREEN)
ax2.set_ylabel('Score', fontsize=12)
ax2.set_ylim(0, 80)
ax2.text(0.5, 0.07,
         f'CIs overlap — difference NOT significant\n'
         f'(A: [{ci_a[0]:.1f}, {ci_a[1]:.1f}]  B: [{ci_b[0]:.1f}, {ci_b[1]:.1f}])',
         ha='center', transform=ax2.transAxes, fontsize=9.5,
         color='#555', style='italic')

plt.suptitle('Error bars prevent false conclusions about group differences',
             fontsize=13, fontweight='bold', color=PRIMARY, y=1.01)
plt.tight_layout()
plt.show()
t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f't-test: t = {t_stat:.3f},  p = {p_val:.3f}  — NOT significant at alpha=0.05')
````

<div style="background-color: #E3F2FD; border-left: 4px solid #1976D2; padding: 15px 20px; margin: 20px 0; border-radius: 5px;"><h3 style="color: #1976D2; margin-top: 0;">Demo 5 — Cherry-Picked Monthly Trend vs Full Year</h3><p style="color: #333; margin-bottom: 0;">Selecting just three months that happen to rise creates a convincing growth narrative. Showing the full 12 months reveals the trend is flat (or even slightly negative).</p></div>

````python
np.random.seed(99)

import calendar
months_full = np.arange(1, 13)
month_names = [calendar.month_abbr[m] for m in months_full]

# Flat trend with noise — but months 4,5,6 happen to rise
revenue = 100 + np.array([2, -5, 1, 3, 8, 12, -4, -7, 2, -3, 1, -2],
                          dtype=float) + np.random.normal(0, 1.5, 12)

cherry = [3, 4, 5]   # 0-indexed: Apr, May, Jun
cherry_names = [month_names[i] for i in cherry]
cherry_vals  = revenue[cherry]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Bad — cherry-picked
ax = axes[0]
ax.plot(cherry_names, cherry_vals, 'o-', color=RED,
        linewidth=2.5, markersize=9)
slope_c, int_c, *_ = stats.linregress([0, 1, 2], cherry_vals)
ax.plot(cherry_names,
        [slope_c * i + int_c for i in range(3)],
        '--', color=ORANGE, linewidth=2, label='"Trend" line')
ax.set_title('"Consistent Monthly Growth!"\n(3 months cherry-picked)',
             fontsize=13, fontweight='bold', color=RED)
ax.set_ylabel('Revenue Index', fontsize=12)
ax.legend(fontsize=10)
ax.set_ylim(90, 125)

# Good — full year
ax2 = axes[1]
ax2.plot(month_names, revenue, 'o-', color=PRIMARY,
         linewidth=2, markersize=7, label='Monthly revenue')
slope_f, int_f, *_ = stats.linregress(months_full, revenue)
ax2.plot(month_names,
         slope_f * months_full + int_f,
         '--', color=TAN, linewidth=2, label=f'Full-year trend (slope={slope_f:.2f})')
for i in cherry:
    ax2.scatter(month_names[i], revenue[i], color=RED, s=120, zorder=5)
ax2.set_title('Full Year: No Real Trend\n(Cherry-picked months highlighted in red)',
              fontsize=13, fontweight='bold', color=GREEN)
ax2.set_ylabel('Revenue Index', fontsize=12)
ax2.tick_params(axis='x', rotation=40)
ax2.legend(fontsize=9)
ax2.set_ylim(85, 125)

plt.suptitle('Cherry-picking creates growth where none exists',
             fontsize=13, fontweight='bold', color=PRIMARY, y=1.01)
plt.tight_layout()
plt.show()
print(f'Full-year trend slope: {slope_f:.3f} — essentially flat.')
````

<div style="background-color: #E3F2FD; border-left: 4px solid #1976D2; padding: 15px 20px; margin: 20px 0; border-radius: 5px;"><h3 style="color: #1976D2; margin-top: 0;">Demo 6 — Fake Trend from a Small Sample (n = 5)</h3><p style="color: #333; margin-bottom: 0;">A regression line through only 5 data points can look compelling in a polished chart. But the confidence band is enormous — reflecting near-total uncertainty about the true relationship. Showing the band honestly communicates how little we know.</p></div>

````python
np.random.seed(5)

n_small = 5
x_s  = np.array([1, 2, 3, 4, 5], dtype=float)
y_s  = x_s * 1.5 + np.random.normal(0, 3, n_small)

slope_s, int_s, r_s, p_s, se_s = stats.linregress(x_s, y_s)

x_pred = np.linspace(0.5, 5.5, 200)
y_pred = slope_s * x_pred + int_s

# Prediction interval
n = len(x_s)
t_crit = stats.t.ppf(0.975, df=n-2)
x_mean = x_s.mean()
SSx    = ((x_s - x_mean)**2).sum()
residuals = y_s - (slope_s * x_s + int_s)
MSE    = (residuals**2).sum() / (n - 2)
se_pred = np.sqrt(MSE * (1/n + (x_pred - x_mean)**2 / SSx))
ci_low  = y_pred - t_crit * se_pred
ci_high = y_pred + t_crit * se_pred

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bad — line only
ax = axes[0]
ax.scatter(x_s, y_s, color=PRIMARY, s=110, zorder=4)
ax.plot(x_pred, y_pred, color=RED, linewidth=2.5,
        label=f'Trend: y = {slope_s:.2f}x + {int_s:.2f}')
ax.set_title('"Clear Upward Trend!"\n(n = 5, no uncertainty shown)',
             fontsize=13, fontweight='bold', color=RED)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.legend(fontsize=10)

# Good — with confidence band
ax2 = axes[1]
ax2.scatter(x_s, y_s, color=PRIMARY, s=110, zorder=4, label='Data (n=5)')
ax2.plot(x_pred, y_pred, color=RED, linewidth=2.5,
         label=f'y = {slope_s:.2f}x + {int_s:.2f}')
ax2.fill_between(x_pred, ci_low, ci_high,
                 alpha=0.25, color=LAVENDER, label='95% prediction interval')
ax2.set_title('Honest: 95% Prediction Interval\nShows Massive Uncertainty',
              fontsize=13, fontweight='bold', color=GREEN)
ax2.set_xlabel('X', fontsize=12)
ax2.set_ylabel('Y', fontsize=12)
ax2.legend(fontsize=9)
ax2.text(0.05, 0.06,
         f'r² = {r_s**2:.2f},  p = {p_s:.3f}  — uncertain with only 5 points',
         transform=ax2.transAxes, fontsize=9.5, color='#555', style='italic')

plt.suptitle('Small samples produce wide uncertainty — never hide it',
             fontsize=13, fontweight='bold', color=PRIMARY, y=1.01)
plt.tight_layout()
plt.show()
print(f'n=5 regression: slope={slope_s:.2f},  r²={r_s**2:.3f},  p={p_s:.3f}')
print('Even a significant p-value here is unreliable with only 5 observations.')
````
